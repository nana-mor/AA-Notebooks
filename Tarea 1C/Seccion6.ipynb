# Ejercicio 6.1-1

Determinación del Número Mínimo y Máximo de Elementos en un Heap de Altura $h$

## Definiciones Clave

### Heap Completo
Un heap es un árbol binario completo, lo que significa que cada nivel, excepto posiblemente el último, está lleno. En el último nivel, los nodos están alineados a la izquierda.

### Altura del Heap ($h$)
La altura de un heap se define como el número de niveles desde la raíz ($h = 0$) hasta el nivel más profundo.

### Número de Nodos por Nivel
Cada nivel $i$ puede contener hasta $2^i$ nodos. La altura $h$ tiene un total de $h + 1$ niveles (incluyendo el nivel 0).

## 1. Número Máximo de Elementos en un Heap de Altura $h$
El número máximo ocurre cuando todos los niveles están completos. La cantidad total de nodos en este caso es la suma de una progresión geométrica:

$$
\text{Máximo de nodos} = 2^0 + 2^1 + 2^2 + \cdots + 2^h
$$

Utilizando la fórmula de la suma de una progresión geométrica:

$$
\text{Máximo de nodos} = 2^{h+1} - 1
$$

## 2. Número Mínimo de Elementos en un Heap de Altura $h$
El número mínimo ocurre cuando el nivel más profundo ($h$) contiene únicamente un nodo. Esto significa que todos los niveles anteriores están llenos. La cantidad total de nodos es:

$$
\text{Mínimo de nodos} = (2^0 + 2^1 + 2^2 + \cdots + 2^{h-1}) + 1
$$

Simplificando:

$$
\text{Mínimo de nodos} = (2^h - 1) + 1 = 2^h
$$

## Conclusión
En un heap de altura $h$:

- **Número mínimo de elementos:** $2^h$
- **Número máximo de elementos:** $2^{h+1} - 1$


# Ejercicio 6.1-2

Altura de un Heap

Para resolver el ejercicio 6.1-2, se pide demostrar que un heap de $n$ elementos tiene una altura de $\lfloor \log n \rfloor$.

## Definición de Heap
Un heap es un árbol binario completo, lo que significa que todos los niveles están completamente llenos, excepto posiblemente el último nivel, que se llena de izquierda a derecha. Este tipo de estructura asegura una buena eficiencia en las operaciones de inserción y eliminación.

## Altura del Heap
La altura de un árbol binario se define como la longitud del camino más largo desde la raíz hasta una hoja. En un heap, la altura depende del número total de elementos $n$.

### Relación entre $n$ y la Altura $h$
**Árbol Binario Completo:**

En un heap completo, cada nivel tiene el doble de nodos que el nivel anterior. El número máximo de nodos en un nivel $i$ es $2^i$. La altura $h$ de un heap es el número de niveles menos uno, y el número total de nodos $n$ en un heap completo de altura $h$ está en el rango:

$$
2^h \leq n < 2^{h+1}
$$

### Derivación de la Altura
Queremos despejar la altura $h$ en función del número de elementos $n$.

Tomando logaritmo base 2 en ambos lados:

$$
h \leq \log_2 n < h + 1
$$

Reorganizando, podemos decir que:

$$
h = \lfloor \log_2 n \rfloor
$$

## Conclusión
Un heap con $n$ elementos tiene una altura de $\lfloor \log_2 n \rfloor$.

Esto significa que la altura crece logarítmicamente en función del número de elementos, lo que asegura una eficiencia de $O(\log n)$ en operaciones de inserción y eliminación, características fundamentales para el buen rendimiento de un heap.


# Ejercicio 6.1-3

Altura de un Árbol Binario de Búsqueda

## Enunciado
Para el ejercicio 6.1-3, se nos pide analizar la altura de un árbol binario de búsqueda (BST, por sus siglas en inglés) en el peor de los casos para $n$ nodos.

## Definición de Árbol Binario de Búsqueda
Un árbol binario de búsqueda es una estructura de datos en forma de árbol en la cual cada nodo tiene como máximo dos hijos. Cumple con la propiedad de que para cada nodo:

- Todos los valores del subárbol izquierdo son menores que el valor del nodo.
- Todos los valores del subárbol derecho son mayores que el valor del nodo.

## Altura del Árbol Binario de Búsqueda
La altura de un árbol binario de búsqueda se define como la longitud del camino más largo desde la raíz hasta una hoja. La eficiencia de varias operaciones (como búsqueda, inserción y eliminación) depende de la altura del árbol.

En el **mejor caso**, el árbol está equilibrado, lo que lleva a una altura del orden de $O(\log n)$, lo cual es muy eficiente.

En el **peor caso**, el árbol está completamente desequilibrado, similar a una lista enlazada. Esto ocurre cuando los nodos se insertan en orden ascendente o descendente, lo que provoca que cada nodo tenga solo un hijo. En este escenario, la altura del árbol será igual al número total de nodos, es decir:

$$
h = n - 1
$$

donde $h$ es la altura del árbol y $n$ es el número de nodos.

Por lo tanto, en el peor caso, la altura de un árbol binario de búsqueda es $O(n)$.

## Conclusión
- En el mejor de los casos, la altura de un árbol binario de búsqueda es $O(\log n)$, lo cual ocurre cuando el árbol está balanceado.
- En el peor de los casos, la altura es $O(n)$, lo cual sucede cuando el árbol está completamente desequilibrado.

La eficiencia de las operaciones en un árbol binario de búsqueda depende de la altura, y para evitar el peor caso, se suelen utilizar variantes de los árboles binarios de búsqueda, como los **árboles balanceados** (AVL o árboles rojo-negro) que garantizan una altura logarítmica en el peor caso.


# Ejercicio 6.1-4

Para resolver el ejercicio 6.1-4, se pregunta dónde podría residir el elemento más pequeño en un max-heap, asumiendo que todos los elementos son distintos.

## Max-Heap

Un max-heap es un tipo de estructura de árbol binario donde el valor de cada nodo es mayor o igual que los valores de sus hijos. Esto implica que el valor máximo del heap siempre está en la raíz (nivel 0).

## Propiedades Importantes del Max-Heap

- Cada nodo tiene un valor mayor o igual que sus hijos.
- El elemento más grande está en la raíz del heap.
- A medida que descendemos en el árbol, los valores disminuyen.

## Ubicación del Elemento Más Pequeño

Dado que el max-heap se construye de forma que los valores disminuyen a medida que se desciende desde la raíz, el elemento más pequeño debe residir en uno de los nodos hoja, ya que todos los nodos en niveles superiores tienen valores más grandes que sus descendientes.

En un max-heap completo, los nodos hoja se encuentran en los niveles más bajos del árbol, específicamente en los niveles $h$ o $h - 1$ (donde $h$ es la altura del heap).

## Conclusión

En un max-heap, el elemento más pequeño se encuentra en uno de los nodos hoja, que están ubicados en los niveles más bajos del árbol (niveles $h$ o $h - 1$). Esto se debe a que en un max-heap, los valores disminuyen a medida que nos alejamos de la raíz, y por lo tanto, los elementos más pequeños deben encontrarse en las hojas.

Para valores de $k \geq 2$, el k-ésimo elemento más grande puede estar en el nivel 1 o en niveles inferiores, dependiendo de la estructura del heap.

## Conclusión

Para un elemento en el índice $i$:

- El hijo izquierdo está en el índice $2i + 1$.
- El hijo derecho está en el índice $2i + 2$.

El arreglo cumple con la propiedad de min-heap si cada elemento es menor o igual que sus hijos (si existen).

En un arreglo en orden ascendente, cada elemento es menor que el siguiente.

# Ejercicio 6.1-8

Para resolver el ejercicio 6.1-8, se pide demostrar que, utilizando la representación de arreglo para almacenar un heap de $n$ elementos, los nodos hoja son los que están indexados por $⌊n/2⌋ + 1, ⌊n/2⌋ + 2, …, n$.

## Representación de un Heap en un Arreglo

Un heap de $n$ elementos se puede representar de forma eficiente utilizando un arreglo. En la representación de un heap como un arreglo, los elementos están organizados de tal manera que:

Para un nodo en la posición $i$ (donde los índices comienzan en 0):
- El hijo izquierdo se encuentra en la posición $2i + 1$.
- El hijo derecho se encuentra en la posición $2i + 2$.
- El padre de un nodo en la posición $i$ (excepto la raíz) se encuentra en la posición $⌊(i - 1)/2⌋$.

## Identificación de los Nodos Hoja

Los nodos hoja en un heap son aquellos que no tienen hijos. En un heap representado en forma de arreglo, esto ocurre para aquellos índices que no cumplen la relación para tener hijos válidos (es decir, los hijos exceden el tamaño del arreglo $n$).

Para un heap de tamaño $n$:
- Los nodos internos (no hojas) son aquellos que tienen al menos un hijo. Los índices de estos nodos van desde $0$ hasta $⌊n/2⌋ - 1$.
- Por lo tanto, los nodos hoja comienzan justo después de $⌊n/2⌋ - 1$, es decir, en la posición $⌊n/2⌋$ y continúan hasta el último nodo en el arreglo.

## Demostración

- **Número de Nodos Internos**: Los nodos internos están indexados desde $0$ hasta $⌊n/2⌋ - 1$.
- **Número de Nodos Hoja**: Los nodos hoja comienzan en el índice $⌊n/2⌋$ y continúan hasta el índice $n - 1$.

En términos de numeración desde $1$, los nodos hoja están indexados por:

$⌊n/2⌋ + 1, ⌊n/2⌋ + 2, …, n$

Esto se debe a que todos los nodos con índices mayores a $⌊n/2⌋$ no tienen hijos, lo que significa que son nodos hoja.

## Conclusión

En la representación de un heap con $n$ elementos usando un arreglo, los nodos hoja son aquellos que están indexados desde:

$⌊n/2⌋ + 1, ⌊n/2⌋ + 2, …, n$

Esto se cumple porque todos los nodos con un índice mayor a $⌊n/2⌋$ ya no tienen hijos y, por lo tanto, son nodos hoja.

# Ejercicio 6.2-2

Mostrar que cada hijo de la raíz de un heap de $n$ nodos es la raíz de un subárbol que contiene a lo más $2n/3$ nodos. ¿Cuál es la constante más pequeña $\alpha$ tal que cada subárbol tiene a lo más $\alpha n$ nodos? ¿Cómo afecta esto la recurrencia (6.1) y su solución?


## Heap de $n$ nodos
Un heap es una estructura de árbol binario completo donde el valor en cada nodo es mayor (en un max-heap) o menor (en un min-heap) que sus hijos.

## División del heap en subárboles
Para un heap de $n$ nodos, la raíz tiene dos hijos, cada uno de los cuales es la raíz de un subárbol. Queremos demostrar que cada uno de estos subárboles tiene, como máximo, $2n/3$ nodos.

La raíz tiene un hijo izquierdo y un hijo derecho. En un árbol binario completo, la raíz tiene a lo sumo dos hijos, y si tiene $n$ nodos, aproximadamente la mitad de los nodos están en el subárbol izquierdo y la otra mitad en el subárbol derecho.

Si consideramos que la raíz tiene dos subárboles, uno con $n_1$ nodos y otro con $n_2$ nodos, entonces $n_1 + n_2 = n - 1$ (porque excluimos la raíz).

## Determinación de $n_1$ y $n_2$
En el peor caso, el subárbol izquierdo o derecho podría contener más nodos que el otro. Supongamos que $n_1 \geq n_2$. Entonces $n_1$ es el subárbol más grande.

Para asegurar que cada subárbol tenga, como máximo, $2n/3$ nodos, necesitamos que:

$$
n_1 \leq \frac{2n}{3}
$$

Dado que $n_1 + n_2 = n - 1$, podemos inferir que:

$$
n_1 \leq \frac{2n}{3}, \quad n_2 \leq \frac{2n}{3}
$$

Esta relación es posible porque estamos dividiendo los nodos de manera equitativa entre los subárboles.

## Constante más pequeña $\alpha$
La constante más pequeña $\alpha$ tal que cada subárbol tiene a lo más $\alpha n$ nodos es $\alpha = \frac{2}{3}$. Esto se debe a que, como hemos mostrado, cada subárbol tiene como máximo $2n/3$ nodos.

## Efecto sobre la recurrencia
La recurrencia (6.1) en el contexto de heaps generalmente describe el tiempo requerido para aplicar una operación (como MAX-HEAPIFY) en un subárbol.

Si cada subárbol tiene como máximo $2n/3$ nodos, la recurrencia se puede expresar como:

$$
T(n) = T\left(\frac{2n}{3}\right) + f(n)
$$

donde $f(n)$ es una función que describe el trabajo hecho fuera de la llamada recursiva, como comparar nodos o intercambiar valores.

Con $\alpha = \frac{2}{3}$, podemos resolver la recurrencia usando el método maestro o sustitución para determinar su complejidad temporal. Por ejemplo, si $f(n)$ es lineal ($f(n) = O(n)$), el tiempo de ejecución seguirá siendo $O(n \log n)$.

En resumen, la constante más pequeña $\alpha$ tal que cada subárbol tiene a lo más $\alpha n$ nodos es $\frac{2}{3}$. Esto implica que cada llamada recursiva afecta a un subárbol que es, como máximo, dos tercios del tamaño del árbol original, lo cual mantiene la recurrencia en $O(n \log n)$ en términos de complejidad temporal para operaciones como HEAPIFY.


# Ejercicio 6.2-3

A partir del procedimiento MAX-HEAPIFY, escribe el pseudocódigo para el procedimiento MIN-HEAPIFY(A, i), que realiza la operación correspondiente.

## Solución

MAX-HEAPIFY asegura que el subárbol con raíz en el índice $i$ cumpla con la propiedad de un max-heap, es decir, que cada nodo sea mayor o igual que sus hijos. De manera similar, MIN-HEAPIFY debe asegurar que el subárbol con raíz en $i$ cumpla con la propiedad de un min-heap, donde cada nodo es menor o igual que sus hijos.

Pseudocódigo para MIN-HEAPIFY
MIN-HEAPIFY(A, i):
    # Encuentra los índices de los hijos izquierdo y derecho
    left = 2 * i
    right = 2 * i + 1
    # Suponemos que el nodo actual (en el índice i) es el más pequeño
    smallest = i

    # Si el hijo izquierdo existe y es menor que el nodo actual, actualiza smallest
    if left <= A.heap_size and A[left] < A[smallest]:
        smallest = left

    # Si el hijo derecho existe y es menor que el nodo más pequeño encontrado, actualiza smallest
    if right <= A.heap_size and A[right] < A[smallest]:
        smallest = right

    # Si el nodo más pequeño no es el nodo actual (i), intercambia y continúa en el subárbol afectado
    if smallest != i:
        swap(A[i], A[smallest])  # Intercambia los valores
        MIN-HEAPIFY(A, smallest)  # Llama recursivamente para arreglar el subárbol
## Explicación del algoritmo
Entrada:
$A$: el arreglo que representa el heap.
$i$: el índice de la raíz del subárbol que queremos ajustar.
Cálculo de hijos: Para un nodo en $i$, su hijo izquierdo está en $2i$ y su hijo derecho en $2i + 1$.
Determinación del más pequeño: Compara el valor del nodo $i$ con sus hijos (si existen) y selecciona el más pequeño.
Intercambio y recursión: Si un hijo es más pequeño que el nodo actual, intercambia el nodo actual con el hijo más pequeño y llama recursivamente a MIN-HEAPIFY sobre el hijo afectado.
Ejemplo de ejecución
Supongamos que tenemos el siguiente arreglo representando un min-heap:

$A = [\infty, 4, 5, 9, 10, 7, 8, 6]$

(La posición 0 no se usa en este ejemplo para simplificar las fórmulas.)

Si llamamos MIN-HEAPIFY(A, 2), el procedimiento hará lo siguiente:
Nodo $i = 2$, valor 5.
Hijo izquierdo: $A[4] = 10$, hijo derecho: $A[5] = 7$.
El menor entre $A[2], A[4], A[5]$ es $A[2] = 5$. No se hace nada porque ya cumple la propiedad del min-heap.
Si llamamos MIN-HEAPIFY(A, 1):
Nodo $i = 1$, valor 4.
Hijo izquierdo: $A[2] = 5$, hijo derecho: $A[3] = 9$.
El menor entre $A[1], A[2], A[3]$ es $A[1] = 4$. No se hace nada porque ya cumple la propiedad del min-heap.
Complejidad
Tiempo: $O(\log n)$, ya que el heap es un árbol binario completo y la altura máxima es $\log n$.
Espacio: $O(1)$, si no contamos el espacio de la pila de recursión.


# Ejercicio 6.2-4
**Pregunta**: ¿Cuál es el efecto de llamar a MAX-HEAPIFY(A, i) cuando el elemento $A[i]$ es mayor que sus hijos?

**Respuesta**:

Cuando se llama a MAX-HEAPIFY(A, i) y el elemento $A[i]$ ya es mayor que sus hijos (es decir, $A[i] \geq A[\text{left}]$ y $A[i] \geq A[\text{right}]$):

### Propiedad del max-heap ya está satisfecha en $i$
- En este caso, no es necesario realizar ningún intercambio o acción adicional, ya que el nodo $i$ cumple con la condición del max-heap.
- Los subárboles con raíz en $left(i)$ y $right(i)$ tampoco serán modificados.

### Efecto práctico
- El procedimiento termina inmediatamente después de verificar que el nodo $A[i]$ es mayor o igual que sus hijos, y no se realizan llamadas recursivas ni intercambios.

### Complejidad
- El caso donde $A[i]$ ya satisface la propiedad del max-heap es el mejor caso para MAX-HEAPIFY.
- La complejidad en este escenario es $O(1)$, ya que únicamente se realizan comparaciones entre $A[i]$ y sus hijos.

### Resumen
- Llamar a MAX-HEAPIFY(A, i) cuando $A[i]$ ya es mayor que sus hijos tiene un efecto trivial, ya que el procedimiento no realiza ninguna modificación ni llamada recursiva adicional, terminando de inmediato. Esto resulta en el mejor caso con un costo constante $O(1)$.


# Problema 6.2-5

Demostrar que el procedimiento BUILD-MAX-HEAP tiene una complejidad temporal de $O(n)$.

## Explicación

BUILD-MAX-HEAP se utiliza para construir un max-heap a partir de un arreglo desordenado. Esto se logra llamando iterativamente a la función MAX-HEAPIFY en todos los nodos que no sean hojas, comenzando desde el último nodo interno hacia la raíz.

El algoritmo de BUILD-MAX-HEAP se puede representar de la siguiente manera:

BUILD-MAX-HEAP(A) for i = floor(length(A) / 2) downto 1 MAX-HEAPIFY(A, i)


Donde MAX-HEAPIFY es el procedimiento que ajusta el subárbol con raíz en el índice $i$ para que satisfaga la propiedad de un max-heap, asumiendo que los subárboles izquierdo y derecho ya cumplen la propiedad de max-heap.

## Complejidad de BUILD-MAX-HEAP

Para demostrar que BUILD-MAX-HEAP tiene una complejidad temporal de $O(n)$, consideremos cuántas veces se llama a MAX-HEAPIFY y cuál es su costo en cada llamada.

En el peor caso, la función MAX-HEAPIFY tiene una complejidad temporal de $O(h)$, donde $h$ es la altura del nodo en el que se aplica. La altura de un nodo $i$ en un árbol binario completo está relacionada con el número de elementos $n$ del árbol. La altura máxima del heap es $\log_2(n)$.

Para analizar la complejidad de BUILD-MAX-HEAP, notemos que el bucle `for` comienza desde el último nodo interno y avanza hacia la raíz del heap. Los nodos más bajos (cerca de las hojas) requieren menos trabajo, mientras que los nodos más altos (cerca de la raíz) requieren más trabajo. Sin embargo, hay más nodos en las partes inferiores del árbol y menos en las superiores.

Podemos analizar el costo total sumando el trabajo realizado en cada nivel. Definamos el trabajo total como:

$$
T(n) = \sum_{i=1}^{\lfloor n/2 \rfloor} O(h_i)
$$

Donde $h_i$ es la altura del nodo $i$. Para simplificar, podemos aproximar la cantidad de trabajo necesario para cada nivel utilizando la siguiente fórmula:

$$
T(n) = \sum_{i=0}^{\lfloor \log_2 n \rfloor} \left( \frac{n}{2^i} \right) O(i)
$$

Esto representa la suma del costo de aplicar MAX-HEAPIFY para todos los nodos en cada nivel del árbol. Sin embargo, esta suma converge a un valor proporcional a $O(n)$. Para demostrar esto, podemos observar que:

- La mayoría de los nodos están cerca de las hojas del árbol, donde el costo de MAX-HEAPIFY es pequeño (casi constante).
- Sólo un número pequeño de nodos está cerca de la raíz, donde el costo de MAX-HEAPIFY es alto.

Si sumamos el trabajo total requerido para cada nodo, obtenemos una complejidad total de $O(n)$ en lugar de $O(n \log n)$, como sería el caso si aplicáramos MAX-HEAPIFY desde la raíz de manera repetida.

## Conclusión

El costo de BUILD-MAX-HEAP se puede demostrar que es $O(n)$ debido a la naturaleza de cómo se propagan los cambios en el heap y el número de nodos en cada nivel. Aunque MAX-HEAPIFY tiene una complejidad de $O(\log n)$, la mayoría de las llamadas en BUILD-MAX-HEAP ocurren en los niveles inferiores, donde el costo es mucho menor, resultando en un costo total de $O(n)$.

```python
def max_heapify_iterative(A, i):
    n = len(A)
    while True:
        left = 2 * i + 1   # índice del hijo izquierdo
        right = 2 * i + 2  # índice del hijo derecho
        largest = i

        # Comparar con el hijo izquierdo
        if left < n and A[left] > A[largest]:
            largest = left

        # Comparar con el hijo derecho
        if right < n and A[right] > A[largest]:
            largest = right

        # Si el nodo actual es más pequeño que alguno de sus hijos, intercambiar
        if largest != i:
            A[i], A[largest] = A[largest], A[i]
            # Continuar desde el nodo que se intercambió
            i = largest
        else:
            # Si no hay más cambios que hacer, salir del bucle
            break
Explicación del Código

1. Inicialización de Variables
left y right se calculan como los índices de los hijos izquierdo y derecho de i, respectivamente.
largest inicialmente se asigna como i, que es el nodo actual.
2. Comparación con los Hijos
Se compara el valor en el índice i con el de sus hijos (left y right).
Si alguno de los hijos tiene un valor mayor, se actualiza largest para que sea el índice del hijo más grande.
3. Intercambio e Iteración
Si largest no es el nodo actual i, se realiza un intercambio entre A[i] y A[largest] para asegurar que el nodo actual cumple con la propiedad de max-heap.
Luego, se actualiza i para continuar iterando desde el nodo que se intercambió (esto asegura que los subárboles también cumplen la propiedad de max-heap).
4. Condición de Parada
Si largest es igual a i, significa que no hay más intercambios necesarios y el bucle termina.

# Ejercicio 6.2-7 

Mostrar que el tiempo de ejecución en el peor caso de MAX-HEAPIFY en un heap de tamaño $n$ es $\Omega(\log n)$. (Pista: Para un heap con $n$ nodos, da valores de nodos que causen que MAX-HEAPIFY sea llamado recursivamente en cada nodo en un camino simple desde la raíz hasta una hoja).
Para demostrar que el tiempo de ejecución en el peor caso de MAX-HEAPIFY es $\Omega(\log n)$, necesitamos encontrar un escenario en el que se realicen la mayor cantidad de llamadas recursivas posibles. Esto ocurrirá cuando MAX-HEAPIFY se aplique desde la raíz y siga propagándose a lo largo del camino hacia una hoja.

Explicación del Problema
MAX-HEAPIFY es un procedimiento que ajusta un subárbol para mantener la propiedad del max-heap. Dado un nodo i, MAX-HEAPIFY compara el valor de i con sus hijos izquierdo y derecho, intercambia con el mayor si es necesario y continúa ajustando desde el nodo hijo afectado.

En el peor caso, MAX-HEAPIFY debe recorrer la altura completa del árbol hasta llegar a una hoja. En un heap binario con n nodos, la altura del árbol es $\log_2(n)$ (aproximadamente). Esto significa que la función debe "bajar" desde la raíz hasta el nodo hoja, lo cual implica un número de pasos proporcional a la altura del árbol.

Ejemplo de Escenario del Peor Caso
Para visualizar un escenario de peor caso, imaginemos un heap binario donde:

El nodo raíz tiene un valor muy pequeño.
Cada nodo interno tiene un valor mayor que el nodo padre pero menor que su hermano derecho.
Por ejemplo, consideremos el siguiente heap (representado como un árbol):

          1
         / \
        3   2
       / \
      5   4
Aquí, si llamamos a MAX-HEAPIFY en la raíz (nodo con valor 1):

Primero comparará 1 con sus hijos 3 y 2. El mayor es 3, así que intercambiamos 1 con 3.
Luego, MAX-HEAPIFY continúa con el nodo que ahora tiene valor 1 (que estaba en la posición del 3).
En la siguiente llamada, 1 se compara con sus hijos 5 y 4. El mayor es 5, por lo que 1 se intercambia con 5.
El procedimiento continúa hasta llegar a una hoja.
En este caso, MAX-HEAPIFY recorre toda la altura del árbol, lo cual corresponde a $\log_2(n)$ pasos.

Complejidad Temporal
En cada nivel del árbol, MAX-HEAPIFY debe comparar el nodo actual con sus hijos y posiblemente hacer un intercambio.
En el peor caso, este proceso se repite desde la raíz hasta una hoja, lo cual implica un número de pasos proporcional a la altura del árbol.
Como la altura de un heap binario con n nodos es $\log_2(n)$, la complejidad temporal en el peor caso de MAX-HEAPIFY es $\Omega(\log n)$.
Por lo tanto, el tiempo de ejecución en el peor caso de MAX-HEAPIFY es $\Omega(\log n)$, ya que en el peor escenario se requiere recorrer toda la altura del árbol, realizando operaciones en cada nivel del mismo.

# Ejercicio 6.3-1
La operación BUILD-MAX-HEAP convierte un arreglo en un "max-heap", donde el valor de cada nodo es mayor o igual al de sus hijos. Para realizar esto, se utiliza la función MAX-HEAPIFY iterativamente comenzando desde la mitad del arreglo hacia atrás (índices que tienen hijos). 

Arreglo Inicial
A = [5, 3, 17, 10, 84, 19, 6, 22, 9]

Paso a Paso de BUILD-MAX-HEAP
Identificación del tamaño del heap:
n = 9 (número de elementos en el arreglo).
Se empieza desde el índice i = ⌊n/2⌋ = ⌊9/2⌋ = 4. Los elementos en índices mayores a 4 son hojas, así que no es necesario "max-heafificarlos".
Aplicación de MAX-HEAPIFY desde i = 4 hasta i = 0:
i = 4 (valor 84):
El nodo 84 ya es mayor que sus hijos (22 y 9), por lo tanto, no se necesita ninguna acción.
i = 3 (valor 10):
Hijos: 22 (izquierdo) y 9 (derecho).
22 es el mayor de los hijos y es mayor que 10, por lo que intercambiamos 10 y 22.
Arreglo después del intercambio: [5, 3, 17, 22, 84, 19, 6, 10, 9].
i = 2 (valor 17):
Hijos: 19 (izquierdo) y 6 (derecho).
19 es el mayor de los hijos y es mayor que 17, por lo que intercambiamos 17 y 19.
Arreglo después del intercambio: [5, 3, 19, 22, 84, 17, 6, 10, 9].
i = 1 (valor 3):
Hijos: 22 (izquierdo) y 84 (derecho).
84 es el mayor de los hijos y es mayor que 3, por lo que intercambiamos 3 y 84.
Arreglo después del intercambio: [5, 84, 19, 22, 3, 17, 6, 10, 9].
Ahora 3 tiene dos hijos (10 y 9). 10 es el mayor y es mayor que 3, por lo que intercambiamos 3 y 10.
Arreglo después del intercambio: [5, 84, 19, 22, 10, 17, 6, 3, 9].
i = 0 (valor 5):
Hijos: 84 (izquierdo) y 19 (derecho).
84 es el mayor de los hijos y es mayor que 5, por lo que intercambiamos 5 y 84.
Arreglo después del intercambio: [84, 5, 19, 22, 10, 17, 6, 3, 9].
Ahora 5 tiene dos hijos (22 y 10). 22 es el mayor y es mayor que 5, por lo que intercambiamos 5 y 22.
Arreglo después del intercambio: [84, 22, 19, 5, 10, 17, 6, 3, 9].
Resultado Final
Después de ejecutar BUILD-MAX-HEAP, el arreglo se convierte en:

A = [84, 22, 19, 5, 10, 17, 6, 3, 9]

Este arreglo ahora representa un max-heap, donde cada nodo es mayor o igual a sus hijos.








# Ejercicio 6.3-2

El ejercicio 6.3-2 pide demostrar que:

$$
\lfloor \frac{n}{2^h} \rfloor + 1 \geq \frac{1}{2} \quad \text{para} \quad 0 \leq h \leq \lfloor \log n \rfloor
$$

## Análisis

- **n**: Número de elementos en el heap.
- **h**: Nivel o altura del nodo dentro del heap.
- **2^h**: Representa la cantidad de nodos que potencialmente pueden existir en el nivel **h** del heap.
- $$ \lfloor \frac{n}{2^h} \rfloor $$: Representa el número de nodos existentes a esa altura del heap después de dividir el número total de elementos (**n**) por el número de nodos posibles en la altura **h**.
- **h** varía entre 0 y $$ \lfloor \log n \rfloor $$, es decir, desde la raíz del heap hasta la altura máxima posible para un heap con **n** elementos.

## Demostración

Queremos probar que:

$$
\lfloor \frac{n}{2^h} \rfloor + 1 \geq \frac{1}{2} \quad \text{para} \quad 0 \leq h \leq \lfloor \log n \rfloor
$$

### Caso base ($h = 0$):

Para **h = 0**, tenemos:

$$
\lfloor \frac{n}{2^0} \rfloor + 1 = \lfloor n \rfloor + 1 = n + 1
$$

**n + 1** siempre será mayor que **1/2** para cualquier valor de **n** positivo.

### Caso general ($0 \leq h \leq \lfloor \log n \rfloor$):

A medida que incrementamos **h**, estamos dividiendo **n** entre una potencia de 2, lo cual disminuye el valor de la expresión.

En el peor de los casos, cuando **h** alcanza su valor máximo posible ($\lfloor \log n \rfloor$), el valor de $$\lfloor \frac{n}{2^h} \rfloor$$ será al menos **1** (ya que estamos considerando que **n** es una potencia de 2 y siempre habrá al menos un nodo en cada nivel). Entonces:

$$
\lfloor \frac{n}{2^h} \rfloor + 1 \geq 1 + 1 = 2
$$

**2** es claramente mayor que **1/2**, por lo que la desigualdad se cumple.

## Conclusión

La desigualdad se cumple para todos los valores de **h** en el rango $$0 \leq h \leq \lfloor \log n \rfloor$$, ya que incluso en el peor de los casos (**h** máximo), el valor de la expresión es al menos **2**, que es mayor que **1/2**.


# Ejercicio 6.3-3

**Pregunta:**

¿Por qué el índice de bucle $i$ en la línea 2 de `BUILD-MAX-HEAP` decrece de $⌊n/2⌋$ a 1 en lugar de incrementarse de 1 a $⌊n/2⌋$?

Para entender esto, debemos tener en cuenta cómo funciona el procedimiento `BUILD-MAX-HEAP`, el cual está diseñado para convertir un arreglo en un max-heap. Este procedimiento se basa en la función `MAX-HEAPIFY` que ajusta los elementos del arreglo de tal manera que satisfagan la propiedad de max-heap (donde el valor de cada nodo es mayor o igual que el valor de sus hijos).

## Justificación

El procedimiento `BUILD-MAX-HEAP` comienza en el índice $⌊n/2⌋$ y va decreciendo hasta 1 (o en algunos textos hasta 0, dependiendo del lenguaje y la notación). Esta elección de comenzar desde $⌊n/2⌋$ y decrementar tiene dos razones fundamentales:

### Eficiencia en la corrección del heap:

1. Los elementos que se encuentran en las posiciones desde $⌊n/2⌋ + 1$ hasta $n$ son hojas del árbol. Estos elementos ya cumplen automáticamente con la propiedad de max-heap porque no tienen hijos.

2. Los nodos que no son hojas (es decir, desde $⌊n/2⌋$ hacia 1) necesitan ser ajustados para garantizar que cumplan con la propiedad de max-heap. Al comenzar desde $⌊n/2⌋$ y descender, estamos asegurándonos de que los subárboles más pequeños se conviertan primero en max-heaps antes de ajustar los nodos superiores. Esto es importante porque cuando se aplica `MAX-HEAPIFY` a un nodo, necesitamos asegurarnos de que sus subárboles ya sean max-heaps para poder propagar correctamente la propiedad del max-heap hacia arriba.

### Construcción de subárboles max-heaps:

`MAX-HEAPIFY` trabaja bajo el supuesto de que los subárboles izquierdo y derecho de un nodo ya son max-heaps, y luego ajusta el nodo para que también cumpla con la propiedad de max-heap.

- Si comenzáramos desde 1 y aumentáramos, los nodos podrían tener hijos que aún no están en la estructura de max-heap, lo cual requeriría múltiples correcciones y haría que el proceso fuera menos eficiente.
- Al proceder de manera descendente desde $⌊n/2⌋$ hasta 1, nos aseguramos de que cada vez que llamemos a `MAX-HEAPIFY` en un nodo, sus subárboles ya cumplan con la propiedad de max-heap, lo cual hace que el proceso de corrección sea correcto y eficiente.

## Ejemplo Intuitivo

Consideremos un heap como un árbol binario. Los nodos más cercanos a las hojas (es decir, aquellos en el nivel más bajo que tienen hijos) deben ser ajustados primero, ya que cualquier ajuste en estos nodos podría afectar la estructura del heap hacia arriba. Ajustando de abajo hacia arriba, garantizamos que cada vez que intentamos ajustar un nodo superior, los subárboles bajo él ya tienen la propiedad de max-heap, haciendo la operación más directa y eficiente.

## Conclusión

El índice $i$ en `BUILD-MAX-HEAP` decrece desde $⌊n/2⌋$ hasta 1 porque:

1. Los elementos desde $⌊n/2⌋ + 1$ hasta $n$ son hojas y ya cumplen la propiedad de max-heap.
2. Decrementar garantiza que los subárboles más pequeños se conviertan en max-heaps antes de ajustar los nodos superiores, asegurando una construcción eficiente del heap completo.
3. `MAX-HEAPIFY` funciona de manera óptima cuando se garantiza que los subárboles ya son max-heaps, lo cual se logra al procesar de abajo hacia arriba.


# Ejercicio 6.3-4

El ejercicio 6.3-4 pide demostrar que:

"En un heap de n elementos, hay como máximo $n/2^h$ + 1 nodos de altura h."

Para resolver esto, es importante tener una buena comprensión de la estructura del heap y de cómo se distribuyen los nodos en cada nivel.

## Conceptos Previos
- Un heap es un árbol binario casi completo que se representa como un arreglo.
- La altura de un nodo es la longitud del camino más largo desde ese nodo hasta una hoja.
- El nivel de un nodo se define como la distancia desde la raíz. La raíz está en el nivel 0, sus hijos en el nivel 1, y así sucesivamente.

Queremos demostrar que el número de nodos de altura h en un heap con n elementos es como máximo $n/2^h$ + 1.

## Explicación
En un heap binario, el número de nodos en el nivel h sigue una estructura que se puede expresar en términos de n.

- Un heap binario de n elementos se puede dividir en niveles. El nivel h contiene, como máximo, $2^h$ nodos.
- Para encontrar cuántos nodos hay en cada nivel, podemos dividir el número total de nodos n por $2^h$.

## Paso a Paso de la Demostración

### Número de Nodos en Cada Altura h
- A medida que descendemos desde la raíz hacia niveles inferiores, el número de nodos se divide aproximadamente a la mitad en cada nivel.
- El número de nodos en la altura h del heap se puede expresar como $n / 2^h$.

### Tomando el Máximo Entero Menor
- Como el número total de nodos n puede no ser exactamente divisible por $2^h$, debemos tomar el máximo entero menor o igual a $n / 2^h$, lo cual se representa con $\lfloor n / 2^h \rfloor$.
- Además, debido a la naturaleza de la división en un árbol binario, el número de nodos en cada altura no será exactamente igual a $n / 2^h$ para todos los valores de h, y puede haber una pequeña variación. Para garantizar que cubrimos todos los nodos, sumamos 1 al resultado.

### Resultado Final
Entonces, el número de nodos de altura h es como máximo $\lfloor n / 2^h \rfloor + 1$.

## Ejemplo
Supongamos un heap con $n = 15$ elementos y queremos analizar cuántos nodos hay en la altura $h = 2$:

- El cálculo sería $n / 2^h = 15 / 2^2 = 15 / 4 = 3.75$.
- Tomando el máximo entero menor o igual ($\lfloor 3.75 \rfloor$), obtenemos 3.
- Sumando 1, tenemos $3 + 1 = 4$.

Esto significa que, en este ejemplo, hay como máximo 4 nodos en la altura $h = 2$.

## Conclusión
En un heap binario con n elementos, hay como máximo $\lfloor n/2^h \rfloor + 1$ nodos de altura h. Esta expresión nos asegura que estamos considerando todas las posibles variaciones en la estructura de los niveles de un heap binario, donde la distribución de los nodos puede no ser uniforme en todos los niveles.


# Ejercicio 6.4-1
Usando la Figura 6.4 como modelo, ilustra la operación de HEAPSORT en el arreglo A = {5, 13, 2, 25, 7, 17, 20, 8, 4}


## Paso 1: Construcción del Max-Heap
El primer paso de HEAPSORT es construir un max-heap a partir del arreglo dado. Un max-heap es una estructura donde cada nodo es mayor o igual que sus hijos.

$A = \{5, 13, 2, 25, 7, 17, 20, 8, 4\}$  
Comenzamos desde el último nodo no hoja y aplicamos "max-heapify" hacia arriba. Al construir el max-heap:

1. Aplicamos max-heapify al nodo con índice 3 (valor 25). No necesita ningún cambio porque ya es mayor que sus hijos.
2. Aplicamos max-heapify al nodo con índice 2 (valor 2). Después de la operación, intercambiamos con su hijo mayor (valor 20).
3. Aplicamos max-heapify al nodo con índice 1 (valor 13). Después de la operación, intercambiamos con su hijo mayor (valor 17).
4. Finalmente, aplicamos max-heapify al nodo raíz (valor 5). Después de la operación, intercambiamos con el valor más grande en el subárbol (valor 25).

Después de construir el max-heap, el arreglo se ve así:

**Max-Heap**: $ \{25, 17, 20, 8, 7, 2, 13, 5, 4\} $

## Paso 2: Ordenamiento
Ahora comenzamos a extraer el elemento máximo (la raíz del heap) y lo colocamos al final del arreglo, reduciendo el tamaño del heap en cada paso.

1. **Extraer el máximo (25)**:
   - Intercambiamos el valor 25 con el último elemento (4).
   - Aplicamos max-heapify para restaurar la propiedad del max-heap.
   - Resultado: $ \{20, 17, 13, 8, 7, 2, 4, 5, 25\} $

2. **Extraer el máximo (20)**:
   - Intercambiamos el valor 20 con el penúltimo elemento (5).
   - Aplicamos max-heapify.
   - Resultado: $ \{17, 8, 13, 5, 7, 2, 4, 20, 25\} $

3. **Extraer el máximo (17)**:
   - Intercambiamos el valor 17 con el antepenúltimo elemento (4).
   - Aplicamos max-heapify.
   - Resultado: $ \{13, 8, 4, 5, 7, 2, 17, 20, 25\} $

4. **Extraer el máximo (13)**:
   - Intercambiamos el valor 13 con el cuarto elemento desde el final (2).
   - Aplicamos max-heapify.
   - Resultado: $ \{8, 7, 4, 5, 2, 13, 17, 20, 25\} $

5. **Extraer el máximo (8)**:
   - Intercambiamos el valor 8 con el quinto elemento desde el final (2).
   - Aplicamos max-heapify.
   - Resultado: $ \{7, 5, 4, 2, 8, 13, 17, 20, 25\} $

6. **Extraer el máximo (7)**:
   - Intercambiamos el valor 7 con el sexto elemento desde el final (2).
   - Aplicamos max-heapify.
   - Resultado: $ \{5, 2, 4, 7, 8, 13, 17, 20, 25\} $

7. **Extraer el máximo (5)**:
   - Intercambiamos el valor 5 con el séptimo elemento desde el final (2).
   - Aplicamos max-heapify.
   - Resultado: $ \{4, 2, 5, 7, 8, 13, 17, 20, 25\} $

8. **Extraer el máximo (4)**:
   - Intercambiamos el valor 4 con el octavo elemento desde el final (2).
   - Resultado: $ \{2, 4, 5, 7, 8, 13, 17, 20, 25\} $

## Resultado Final
El arreglo ordenado después de aplicar HEAPSORT es:

$ \{2, 4, 5, 7, 8, 13, 17, 20, 25\} $

Esto ilustra el proceso de HEAPSORT, que comienza construyendo un max-heap y luego extrae repetidamente el elemento máximo para formar el arreglo ordenado.


# Ejercicio 6.4-2
Argumenta la corrección de HEAPSORT usando el siguiente invariante del bucle

Invariante del bucle: Al inicio de cada iteración del bucle for de las líneas 2-5,

El subarreglo $A[1 : i]$ es un max-heap que contiene los $i$ elementos más pequeños de $A[1 : n]$.

El subarreglo $A[i + 1 : n]$ contiene los $n - i$ elementos más grandes de $A[1 : n]$, ordenados.

Vamos a explicar cómo este invariante es mantenido y cómo garantiza la corrección de HEAPSORT.

1. Invariante Inicial:

Al inicio del algoritmo, HEAPSORT construye un max-heap a partir del arreglo $A[1 : n]$. En este punto, el invariante está claramente satisfecho:

El max-heap completo contiene todos los elementos del arreglo original, y está representado por el subarreglo $A[1 : n]$.
No se han ordenado aún los elementos, así que no hay elementos que pertenezcan a la parte ordenada $A[i + 1 : n]$.

2. Mantenimiento del Invariante:

Durante el bucle for, HEAPSORT procede de la siguiente manera:

Extracción del Elemento Máximo: En cada iteración del bucle for, el elemento máximo (que está en la raíz del heap) se intercambia con el último elemento del subarreglo no ordenado. Esto coloca el elemento más grande en la posición correcta en la parte ordenada del arreglo ($A[i + 1 : n]$).

Reducción del Heap y Aplicación de Max-Heapify: Luego, el tamaño del heap se reduce en 1 y se aplica max-heapify al nuevo subarreglo $A[1 : i]$ para restaurar la propiedad del max-heap. Después de aplicar max-heapify, el subarreglo $A[1 : i]$ es un max-heap que contiene los $i$ elementos más pequeños del arreglo original.

Ordenación de los Elementos Mayores: Los elementos más grandes se van colocando en el subarreglo $A[i + 1 : n]$ en cada iteración. Este subarreglo permanece ordenado porque el mayor elemento del heap es extraído y colocado en la siguiente posición disponible al final.
El invariante se mantiene en cada iteración porque:

El subarreglo $A[1 : i]$ sigue siendo un max-heap después de aplicar max-heapify.

El subarreglo $A[i + 1 : n]$ siempre contiene los elementos más grandes, colocados en orden descendente, porque cada extracción coloca el máximo en la posición adecuada.

3. Terminación del Algoritmo:

Cuando el bucle for termina, todos los elementos del arreglo están en el subarreglo $A[i + 1 : n]$, y el subarreglo $A[1 : i]$ está vacío. Dado que cada elemento se ha colocado en su posición correcta a lo largo del proceso, el arreglo completo $A[1 : n]$ está ordenado de manera ascendente.

Conclusión:

El invariante del bucle garantiza que:

En cada iteración, los elementos más grandes del arreglo se extraen y colocan en la posición correcta.
Los elementos restantes siguen siendo un max-heap, lo cual permite que siempre podamos extraer el siguiente elemento más grande.
De este modo, el invariante demuestra la corrección del algoritmo HEAPSORT, asegurando que el arreglo se ordena correctamente al final del proceso.



# Ejercicio 6.4-3
Tiempo de Ejecución de HEAPSORT en un Arreglo Ordenado

Pregunta: ¿Cuál es el tiempo de ejecución de HEAPSORT en un arreglo A de longitud n que ya está ordenado en orden creciente? ¿Qué ocurre si el arreglo ya está ordenado en orden decreciente?

Para responder a esto, consideraremos el comportamiento del algoritmo HEAPSORT en función del estado inicial del arreglo.

Caso 1: Arreglo Ordenado en Orden Creciente

Cuando el arreglo ya está ordenado en orden creciente, HEAPSORT sigue los mismos pasos estándar:

Construcción del Max-Heap:El primer paso de HEAPSORT es construir un max-heap a partir del arreglo dado. La construcción del max-heap requiere un tiempo de $O(n)$, ya que el proceso de max-heapify se aplica en cada nodo comenzando desde los nodos no hoja hacia arriba.Incluso si el arreglo ya está ordenado en orden creciente, el proceso de construcción del max-heap toma tiempo lineal $O(n)$, ya que todos los elementos deben reorganizarse para formar un max-heap.

Extracción de los Elementos y Ordenamiento:Después de construir el max-heap, el algoritmo extrae el elemento máximo repetidamente y lo coloca en su posición correcta.Cada extracción del elemento máximo toma tiempo $O(\log n)$, y este proceso se repite $n$ veces. Por lo tanto, el tiempo total para esta parte es $O(n \log n)$.

En conclusión, el tiempo de ejecución de HEAPSORT para un arreglo ya ordenado en orden creciente sigue siendo:



Esto se debe a que el algoritmo no tiene ninguna optimización especial para detectar si el arreglo ya está ordenado, y procede con los mismos pasos de construcción del heap y extracción de elementos.

Caso 2: Arreglo Ordenado en Orden Decreciente

Si el arreglo ya está ordenado en orden decreciente, HEAPSORT también seguirá los mismos pasos estándar:

Construcción del Max-Heap:Al construir el max-heap, en este caso, el arreglo ya está en una disposición que se aproxima al formato requerido para un max-heap, dado que está en orden decreciente.Aun así, la construcción del max-heap toma tiempo $O(n)$, ya que max-heapify se aplica a todos los nodos desde los nodos no hoja hacia arriba.

Extracción de los Elementos y Ordenamiento:Similar al caso anterior, después de construir el max-heap, se extrae el elemento máximo repetidamente.Cada extracción toma $O(\log n)$ y esto se repite $n$ veces, dando un tiempo total de $O(n \log n)$.

En resumen, el tiempo de ejecución de HEAPSORT para un arreglo ordenado en orden decreciente también es:



Conclusión:

El tiempo de ejecución de HEAPSORT es $O(n \log n)$ en ambos casos, ya sea que el arreglo esté ordenado en orden creciente o decreciente. Esto se debe a que el algoritmo no hace suposiciones sobre el orden inicial del arreglo y siempre realiza la construcción del max-heap y las extracciones de manera consistente.

HEAPSORT tiene una complejidad de tiempo de $O(n \log n)$ en el peor, mejor y caso promedio debido a la necesidad de mantener la estructura del max-heap durante todo el proceso de ordenación, independientemente del estado inicial del arreglo.

# Ejercicio 6.4-4

Demuestra que el tiempo de ejecución en el peor caso de HEAPSORT es $Ω(n \log n)$

Para demostrar que el tiempo de ejecución en el peor caso de HEAPSORT es $Ω(n \log n)$, podemos analizar cómo funcionan las dos fases principales del algoritmo:

1. Construcción del Max-Heap.
2. Extracción repetida del elemento máximo y reestructuración del heap.

Vamos a justificar por qué cada una de estas fases contribuye a la complejidad de tiempo $Ω(n \log n)$.

---

### Fase 1: Construcción del Max-Heap

La construcción del max-heap a partir de un arreglo desordenado implica aplicar el proceso **max-heapify** desde el último nodo no hoja hasta la raíz.

- El proceso de **max-heapify**, en el peor caso, tiene una complejidad de tiempo de $O(\log n)$ para cada nodo.
- Sin embargo, debido a que los nodos más cercanos a la parte inferior del árbol tienen menos niveles por recorrer, la construcción del max-heap tiene una complejidad total de $O(n)$. En términos de notación asintótica, la complejidad para la construcción del heap es $Θ(n)$.

Aunque la construcción del heap tiene una complejidad lineal, $Θ(n)$, esto no afecta el orden de magnitud de la complejidad total del algoritmo.

---

### Fase 2: Extracción del Máximo y Reestructuración

Luego de construir el max-heap, el siguiente paso es extraer el elemento máximo repetidamente y restaurar la propiedad del heap.

- En cada iteración, extraemos el elemento máximo (la raíz del heap) y lo intercambiamos con el último elemento del subarreglo aún no ordenado.
- Después de cada extracción, se debe aplicar **max-heapify** para restaurar la estructura del max-heap. **max-heapify** tiene una complejidad de tiempo de $O(\log n)$.
- Este proceso se repite $n$ veces, ya que tenemos que ordenar $n$ elementos.

Por lo tanto, la complejidad de tiempo para la fase de extracción es:

$$
n \times O(\log n) = O(n \log n)
$$

---

### Complejidad en el Peor Caso

- La fase de construcción del max-heap tiene una complejidad de $Θ(n)$.
- La fase de extracción tiene una complejidad de $O(n \log n)$.
- La fase que domina el tiempo de ejecución total es la fase de extracción, que tiene una complejidad de $O(n \log n)$.

Por lo tanto, el tiempo de ejecución en el peor caso de HEAPSORT es al menos $Ω(n \log n)$.

---

### Demostración Formal

En la notación $Ω$, queremos demostrar una cota inferior para el tiempo de ejecución del algoritmo.

- En el peor caso, cada extracción del elemento máximo requiere que se reconstruya el heap, lo cual lleva un tiempo proporcional a $O(\log n)$ para cada extracción.
- Dado que se realizan $n$ extracciones, el tiempo total es al menos:

$$
Ω(n \log n)
$$

Esto significa que el tiempo de ejecución del peor caso de HEAPSORT no puede ser mejor que $Ω(n \log n)$. La naturaleza de la extracción y reestructuración del heap asegura que este límite inferior siempre se cumpla.

---

### Conclusión

El tiempo de ejecución en el peor caso de HEAPSORT es $Ω(n \log n)$. Esto se debe a que, después de construir el max-heap, cada extracción del elemento máximo implica una operación de **max-heapify** con una complejidad de $Ω(\log n)$, y este proceso se repite $n$ veces, lo que lleva a una complejidad total de $Ω(n \log n)$.


# Ejercicio 6.4-5

Muestra que cuando todos los elementos de $A$ son distintos, el tiempo de ejecución en el mejor caso de HEAPSORT es $\Omega(n \log n)$.


### Análisis del Mejor Caso de HEAPSORT
El algoritmo HEAPSORT consta de dos fases principales:

1. Construcción del Max-Heap.
2. Extracción repetida del elemento máximo y reestructuración del heap.



#### Fase 1: Construcción del Max-Heap
En esta fase, se transforma el arreglo dado en un max-heap. El proceso consiste en aplicar la operación max-heapify desde el último nodo no hoja hasta la raíz del árbol.

- En el mejor caso, el arreglo podría ya estar en una forma que se asemeje a un max-heap o tenga algunos elementos en posiciones favorables, lo cual podría minimizar el número de intercambios.
- Aun en el mejor caso, el proceso de construcción del heap requiere tiempo $\Theta(n)$. Esto se debe a que necesitamos asegurarnos de que cada subárbol cumple la propiedad de heap.

#### Fase 2: Extracción del Máximo y Restauración del Heap
Una vez construido el max-heap, comenzamos a extraer el elemento máximo (es decir, la raíz) y lo colocamos en la última posición del subarreglo no ordenado.

- Después de extraer el elemento máximo, necesitamos restaurar la propiedad del max-heap con max-heapify, que tiene una complejidad de tiempo de $O(\log n)$ para cada extracción.
- Este proceso de extracción se repite $n$ veces, ya que necesitamos ordenar todos los elementos.

En el mejor caso, el proceso de extracción del máximo y restauración del heap no se puede evitar, ya que cada extracción requiere que la estructura del max-heap se mantenga. Esto implica que:

$$n \times O(\log n) = O(n \log n)$$

### Argumento del Mejor Caso
- Incluso si el arreglo original está parcialmente ordenado o tiene propiedades favorables, el algoritmo HEAPSORT siempre necesita construir el max-heap y luego extraer los elementos de manera sistemática.
- Cada extracción del máximo y el consiguiente max-heapify requiere tiempo $\Omega(\log n)$, y este proceso se repite $n$ veces.

Por lo tanto, en el mejor de los casos, el tiempo total de ejecución sigue siendo $\Omega(n \log n)$.

### Conclusión
El tiempo de ejecución en el mejor caso de HEAPSORT, cuando todos los elementos son distintos, es $\Omega(n \log n)$. Esto se debe a que, para ordenar el arreglo, es necesario construir el max-heap y luego extraer los elementos uno a uno, aplicando max-heapify cada vez. Esta operación, que toma tiempo $\Omega(\log n)$, se realiza $n$ veces, lo que lleva a una complejidad de $\Omega(n \log n)$.

HEAPSORT no tiene una mejora significativa en el mejor caso debido a la naturaleza de la construcción y mantenimiento del heap. Por lo tanto, tanto el mejor, peor, y caso promedio tienen una complejidad asintótica de $\Theta(n \log n)$.
