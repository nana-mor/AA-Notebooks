Cuando todos los elementos del subarreglo $A[p:r]$ tienen el mismo valor, el algoritmo PARTITION normalmente colocará el pivote en la posición final después de iterar sobre todo el arreglo, ya que todos los elementos son menores o iguales al pivote.

- Como todos los elementos son iguales, PARTITION finalmente colocará el pivote en la posición $r$, y $q$ será igual al índice del último elemento del subarreglo.

Por lo tanto, para un subarreglo con todos los valores iguales, el valor $q$ retornado por PARTITION será el índice del pivote, usualmente $r$ (el último elemento).

- Si todos son iguales, colocamos el pivote en la posición $(p + r) / 2$.

```python
def modified_partition(A, p, r):
    # Asumimos que el pivote es el último elemento
    pivot = A[r]
    all_equal = True

    # Verificamos si todos los elementos son iguales al pivote
    for j in range(p, r):
        if A[j] != pivot:
            all_equal = False
            break

    # Si todos los elementos son iguales, colocamos el pivote en la posición media
    if all_equal:
        q = (p + r) // 2
        A[q], A[r] = A[r], A[q]  # Intercambiamos el pivote con el elemento en la posición media
        return q

    # Algoritmo estándar de PARTITION si no todos los elementos son iguales
    i = p - 1
    for j in range(p, r):
        if A[j] <= pivot:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i + 1], A[r] = A[r], A[i + 1]
    return i + 1


# Ejercicio 7.1-3

$Para resolver este problema, debemos analizar el tiempo de ejecución del procedimiento PARTITION en un subarreglo de tamaño n.$

Análisis del Tiempo de Ejecución del Algoritmo PARTITION

El procedimiento PARTITION toma como entrada un subarreglo $A[p:r]$ y selecciona un pivote (generalmente el último elemento del subarreglo) para reorganizar los elementos de manera que:

Todos los elementos menores o iguales al pivote estén a la izquierda.
    
Todos los elementos mayores al pivote estén a la derecha.
    
Para lograr esta división, PARTITION utiliza dos índices: $i$ y $j$, y recorre el subarreglo una vez, comparando cada elemento con el pivote.


Inicialización de $i$: El índice $i$ se inicializa con $p - 1$. Esta es una operación constante, $O(1)$.

Iteración del Índice $j$: El índice $j$ recorre desde $p$ hasta $r - 1$, comparando cada elemento con el pivote. Esto implica iterar sobre los $n$ elementos del subarreglo.

Intercambios y Comparaciones: Por cada elemento $A[j]$ en el subarreglo, se realiza una comparación con el pivote. Si $A[j]$ es menor o igual al pivote, se incrementa $i$ y se realiza un intercambio entre $A[i]$ y $A[j]$. Tanto las comparaciones como los intercambios son operaciones de tiempo constante, $O(1)$.

En el peor caso, PARTITION recorre todos los elementos del subarreglo una vez, realizando una comparación por cada elemento. Por lo tanto, el tiempo total para recorrer los elementos es $O(n)$.

Colocación del Pivote: Finalmente, después de completar el bucle, el pivote se coloca en su posición final correcta ($A[i + 1]$). Esta operación de intercambio es también de tiempo constante, $O(1)$.

Tiempo de Ejecución Total

La iteración sobre el subarreglo tiene un costo de $O(n)$, donde $n$ es el tamaño del subarreglo.

Las demás operaciones dentro del bucle (comparaciones e intercambios) son todas de tiempo constante.

La colocación del pivote también toma tiempo constante.

Por lo tanto, el tiempo total de ejecución del procedimiento PARTITION es proporcional al tamaño del subarreglo, es decir, $\Theta(n)$.

Conclusión
El tiempo de ejecución de PARTITION en un subarreglo de tamaño $n$ es $\Theta(n)$, ya que cada elemento del subarreglo es procesado una sola vez, y todas las operaciones realizadas durante el recorrido son de tiempo constante. Esto significa que el tiempo de ejecución crece linealmente con el tamaño del subarreglo, cumpliendo con la notación $\Theta(n)$.

# Ejercicio 7.1-4

Probar que el peor caso de tiempo de ejecución para QUICKSORT es $\Theta(n^2)$

## Planteamiento del Problema

QUICKSORT es un algoritmo de ordenamiento que, en promedio, tiene un tiempo de ejecución de $\Theta(n \log n)$. Sin embargo, su peor caso es conocido por ser $\Theta(n^2)$. El problema 7.1-4 requiere demostrar que el peor caso de tiempo de ejecución para QUICKSORT es, de hecho, $\Theta(n^2)$.

## Análisis del Algoritmo QUICKSORT

El algoritmo QUICKSORT funciona de la siguiente manera:

1. **PARTITION**: Divide el arreglo en dos subarreglos, colocando todos los elementos menores al pivote a la izquierda y todos los elementos mayores a la derecha.
2. **Recursión**: Luego se aplica QUICKSORT de forma recursiva en los dos subarreglos generados.

## Peor Caso para QUICKSORT

El peor caso para QUICKSORT ocurre cuando el procedimiento PARTITION selecciona un pivote que es siempre el elemento más grande o el más pequeño del subarreglo. Esto sucede típicamente cuando el arreglo está ordenado o invertido (es decir, en orden descendente).

Consideremos el proceso en el peor caso:

- **Selección del Pivote**: Si el pivote es siempre el elemento más pequeño o más grande, uno de los subarreglos siempre tendrá tamaño $n-1$, y el otro tendrá tamaño $0$. Esta situación provoca que no haya balance en la división del arreglo.
- **Número de Llamadas Recursivas**: En cada nivel de recursión, el tamaño del subarreglo se reduce en solo uno (el tamaño del problema pasa de $n$ a $n-1$, luego a $n-2$, y así sucesivamente). En este caso, se realizan $n$ niveles de recursión.
- **Costo en Cada Nivel**: En cada nivel de recursión, el costo de particionar el subarreglo es $O(n)$. Esto es porque PARTITION recorre todos los elementos del subarreglo para colocar el pivote en su posición correcta.

## Cálculo del Tiempo de Ejecución

Para entender el tiempo de ejecución total, podemos sumar el costo de cada nivel de recursión:

- **Costo Total**: En cada uno de los $n$ niveles de recursión, el costo es $O(n)$. Por lo tanto, el tiempo total de ejecución es:

  \[
  T(n) = n + (n-1) + (n-2) + \dots + 1 = \frac{n(n+1)}{2}
  \]

  Esto equivale a $O(n^2)$.

- **Notación Theta**: También podemos concluir que el tiempo de ejecución es $\Theta(n^2)$ en el peor caso, ya que el costo crece cuadráticamente con respecto al tamaño del arreglo.

## Conclusión

El peor caso de tiempo de ejecución para QUICKSORT es $\Theta(n^2)$ debido a la falta de balance en la división de los subarreglos cuando el pivote elegido es el elemento más grande o más pequeño. En este escenario, el costo en cada nivel de recursión es lineal, y el número de niveles de recursión también es lineal, lo cual da como resultado un tiempo de ejecución total de $\Theta(n^2)$.


# Ejercicio 7.2-1 

usando el método de sustitución:

Tenemos la recurrencia:

$T(n) = T(n - 1) + Θ(n)$

Queremos demostrar que la solución es $T(n) = Θ(n^2)$.

### Paso 1: Hacer una conjetura
Proponemos que $T(n) ≤ c n^2$ para un valor positivo constante $c$. Esta es una conjetura que verificaremos por inducción.

### Paso 2: Base de la inducción
Para $n = 1$:

$T(1) = Θ(1)$

Podemos asumir que $T(1) ≤ c$, lo cual es cierto si elegimos $c$ adecuadamente grande.

### Paso 3: Paso inductivo
Asumimos que $T(k) ≤ c k^2$ es cierto para algún $k$. Debemos probar que $T(k + 1) ≤ c (k + 1)^2$.

Usando la recurrencia:

$T(k + 1) = T(k) + Θ(k + 1)$

Por hipótesis inductiva:

$T(k) ≤ c k^2$

Por lo tanto:

$T(k + 1) ≤ c k^2 + Θ(k + 1)$

Queremos probar que:

$c k^2 + Θ(k + 1) ≤ c (k + 1)^2$

Expandiendo $(k + 1)^2$:

$(k + 1)^2 = k^2 + 2k + 1$

Por lo tanto:

$c(k^2) + Θ(k + 1) ≤ c(k^2 + 2k + 1) = c k^2 + 2c k + c$

Podemos ver que si $Θ(k + 1) ≤ 2c k + c$, la desigualdad se cumple para un valor adecuado de la constante $c$.

### Conclusión
Hemos demostrado por inducción que:

$T(n) = Θ(n^2)$

por lo tanto, la solución de la recurrencia es $T(n) = Θ(n^2)$.


#Ejercicio 7.2-2
## Análisis del Caso

Si todos los elementos del arreglo $A$ son iguales, QUICKSORT se comportará de manera especial:

- En cada llamada a PARTITION, los elementos serán comparados con el pivote, pero como todos los elementos son iguales, no se producirán cambios significativos en el arreglo.
- Los subarreglos resultantes después de cada partición estarán muy desbalanceados: uno de ellos tendrá todos los elementos menos uno, y el otro estará vacío.

A continuación, analizaremos el tiempo de ejecución:

## Partición de Elementos Iguales
- PARTITION dividirá el arreglo en dos partes: una parte vacía y una parte que contiene todos los elementos menos uno.
- Este comportamiento se repetirá en cada llamada recursiva, de manera que el problema se reducirá a particionar un arreglo de tamaño $n-1$, luego uno de tamaño $n-2$, y así sucesivamente.

## Número de Comparaciones
- En cada llamada a PARTITION, se realizan aproximadamente $n-1$ comparaciones para colocar el pivote.
- Luego, se hace una llamada recursiva sobre un subarreglo de tamaño $n-1$.
- Esto da lugar a una recurrencia similar a la de BUBBLE SORT o INSERTION SORT en el peor caso:
  $T(n) = T(n-1) + Θ(n)$

## Resolución de la Recurrencia
- La solución de esta recurrencia es $T(n) = Θ(n^2)$.

## Conclusión
Cuando todos los elementos del arreglo $A$ tienen el mismo valor, el tiempo de ejecución de QUICKSORT es $Θ(n^2)$.

Esto se debe a que la partición siempre produce un subarreglo vacío y un subarreglo de tamaño $n-1$, lo que resulta en un comportamiento recursivo similar al peor caso de QUICKSORT, donde el pivote está mal balanceado en cada partición.


# Ejercicio 7.2-3
El problema 7.2-3 nos pide demostrar que el tiempo de ejecución de QUICKSORT es $\Theta(n^2)$ cuando el arreglo $A$ contiene elementos distintos y está ordenado en orden decreciente.

Análisis del Caso

Consideremos el comportamiento de QUICKSORT cuando el arreglo está ordenado en orden decreciente, y el algoritmo siempre elige el último elemento del subarreglo como el pivote (un pivote "malo" en este caso).

Selección del Pivote

Si el arreglo está en orden decreciente, cada vez que el último elemento se selecciona como pivote, resulta ser el menor elemento del subarreglo. Esto significa que el subarreglo se dividirá en dos partes: una parte vacía y otra que contiene todos los elementos restantes, menos uno.

Desbalance de la Partición

En cada paso de partición, uno de los subarreglos estará vacío, mientras que el otro tendrá $n - 1$ elementos. Esto resulta en un desbalance extremo, similar al peor caso de QUICKSORT, en el cual el tiempo de ejecución no se beneficia de la división en mitades, sino que se acerca al comportamiento lineal de recorrer el arreglo completo en cada paso.

Recurrencia del Tiempo de Ejecución

Podemos definir la recurrencia para el tiempo de ejecución como:



donde $\Theta(n)$ corresponde al trabajo necesario para la partición, que implica recorrer todo el subarreglo.

Resolución de la Recurrencia

La solución de la recurrencia es $T(n) = \Theta(n^2)$. Esto se debe a que, en cada nivel de recursión, el tamaño del problema se reduce solo en una unidad, pero cada llamada implica un recorrido completo del subarreglo restante.

Conclusión

Cuando el arreglo $A$ contiene elementos distintos y está ordenado en decreciente, el tiempo de ejecución de QUICKSORT es $\Theta(n^2)$.

Este es un peor caso para QUICKSORT, ya que cada partición produce un desbalance máximo, resultando en $n$ niveles de recursión, y cada nivel toma tiempo lineal $\Theta(n)$ para recorrer el subarreglo.



# Ejercicio 7.2-4

El problema 7.2-4 pregunta por qué el algoritmo INSERTION-SORT podría ser más eficiente que QUICKSORT al convertir transacciones bancarias ordenadas por tiempo de transacción a un orden por número de cheque, donde los números de cheque se escriben en orden ascendente, pero los registros están ordenados por tiempo de transacción.

Análisis del Caso

La situación descrita involucra una lista de transacciones bancarias ordenadas según el tiempo de transacción. Queremos reorganizar esta lista para que esté en orden ascendente según el número de cheque. Sin embargo, dado que los cheques suelen ser escritos y procesados en un orden bastante cercano al orden numérico, podemos asumir que la lista inicial está casi ordenada.

Los algoritmos en este contexto:

INSERTION-SORT
INSERTION-SORT es un algoritmo eficiente cuando la lista está casi ordenada. Su complejidad en el mejor caso es $Θ(n)$, cuando la lista ya está ordenada o casi ordenada.

En cada paso, INSERTION-SORT compara un elemento con los elementos anteriores y lo coloca en la posición correcta. Dado que la lista está casi ordenada (es decir, pocos elementos están fuera de lugar), el número de movimientos y comparaciones es relativamente pequeño.

En el peor caso (cuando la lista está ordenada de manera inversa), INSERTION-SORT tiene una complejidad de $Θ(n^2)$, pero este no es el caso aquí.

QUICKSORT
QUICKSORT tiene una complejidad esperada de $Θ(n \log n)$, pero en el peor caso puede ser $Θ(n^2)$ si el pivote elegido no divide bien el arreglo.

Incluso en el caso promedio, QUICKSORT realiza más comparaciones y movimientos de elementos que INSERTION-SORT cuando la lista está casi ordenada.

Comparación y Explicación

Dado que la lista de transacciones está casi ordenada por el número de cheque, INSERTION-SORT es más adecuado porque aprovecha esta característica, moviendo solo unos pocos elementos y haciendo muy pocas comparaciones.

QUICKSORT, por otro lado, no tiene ninguna ventaja especial para listas casi ordenadas, y aún tendrá que dividir la lista en particiones y realizar múltiples comparaciones, lo cual no es tan eficiente como INSERTION-SORT en este caso.

Conclusión

La razón por la cual INSERTION-SORT podría vencer a QUICKSORT para convertir el orden de las transacciones bancarias en este problema es que INSERTION-SORT es particularmente eficiente para listas casi ordenadas, con una complejidad de $Θ(n)$ en el mejor caso. Dado que los cheques suelen ser escritos y procesados en un orden cercano al orden numérico, la lista inicial probablemente estará casi ordenada, haciendo que INSERTION-SORT sea la opción más rápida y eficiente.

# Ejercicio 7.2-5
# Análisis del Árbol de Recursión

## Profundidad mínima

Corresponde al caso en el que siempre se escoge la parte más pequeña para continuar dividiendo. En este caso, estamos dividiendo repetidamente el tamaño del problema por $\alpha$. La profundidad mínima del árbol se puede encontrar resolviendo la siguiente ecuación:

$$n \cdot \alpha^d = 1$$

Donde $n$ es el tamaño original del problema y $d$ es la profundidad. Tomando logaritmo en ambos lados:

$$\log_{\alpha}(n) = d$$

Como $\log_{\alpha} (n) = \frac{\log (n)}{\log (\alpha)}$, la profundidad mínima es:

$$d_{\text{mín}} = \log_{1/\alpha} (n) = \frac{\log (n)}{\log (1/\alpha)} = \log_{1/\alpha} (n)$$

La profundidad mínima es aproximadamente $\log_{1/\alpha} n$.

## Profundidad máxima

Corresponde al caso en el que siempre se escoge la parte más grande para continuar dividiendo. En este caso, estamos dividiendo repetidamente el tamaño del problema por $\beta$. La profundidad máxima del árbol se puede encontrar resolviendo la siguiente ecuación:

$$n \cdot \beta^d = 1$$

Tomando logaritmo en ambos lados:

$$\log_{\beta}(n) = d$$

La profundidad máxima es:

$$d_{\text{máx}} = \log_{1/\beta} (n) = \frac{\log (n)}{\log (1/\beta)} = \log_{1/\beta} (n)$$

La profundidad máxima es aproximadamente $\log_{1/\beta} n$.

## Conclusión

La profundidad mínima de una hoja en el árbol de recursión es aproximadamente $\log_{1/\alpha} n$, y la profundidad máxima es aproximadamente $\log_{1/\beta} n$. Esto se debe a cómo se divide el problema en cada paso, dependiendo de la proporción de $\alpha$ a $\beta$ entre las particiones.


# Ejercicio 7.2-6

## Contexto
Para resolver el problema 7.2-6, necesitamos analizar la probabilidad de que la operación PARTITION en Quicksort produzca una división al menos tan equilibrada como $(1-\alpha)$ a $\alpha$, donde $0 < \alpha \leq \frac{1}{2}$. El objetivo es demostrar que la probabilidad de obtener una división tan equilibrada es aproximadamente $1 - 2\alpha$.

### Contexto del Problema
- Se nos da un arreglo con elementos distintos.
- Todas las permutaciones de los elementos son igualmente probables.
- Se utiliza un procedimiento de particionado (PARTITION) que selecciona un pivote y divide el arreglo en dos subarreglos: los elementos menores que el pivote y los elementos mayores.

### Condición de Equilibrio
Queremos saber cuán probable es que la división sea al menos tan equilibrada como $(1-\alpha)$ a $\alpha$. Esto significa que la partición debe dividir el arreglo de tal forma que el tamaño de los subarreglos esté dentro de la proporción $(1-\alpha) : \alpha$.

En términos más prácticos, queremos que el pivote divida el arreglo de modo que no queden menos de $\alpha n$ elementos en un subarreglo y más de $(1-\alpha) n$ elementos en el otro.

## Derivación de la Probabilidad
Consideremos un arreglo de tamaño $n$ con elementos distintos. La operación PARTITION elige un pivote de manera aleatoria y divide el arreglo en dos subarreglos.

Para que la división sea al menos tan equilibrada como $(1-\alpha)$ a $\alpha$, necesitamos evitar que el pivote se ubique en las posiciones extremas que provoquen un desbalance significativo.

- La condición para tener una partición balanceada es que el pivote no esté en las primeras $\alpha n$ posiciones ni en las últimas $\alpha n$ posiciones del arreglo. Esto asegura que ambas particiones sean de al menos $\alpha n$ elementos.
- La cantidad de posiciones donde el pivote no debe estar es $2\alpha n$ (es decir, las primeras $\alpha n$ y las últimas $\alpha n$).
- La cantidad de posiciones donde el pivote puede estar para producir una partición al menos tan equilibrada es $(n - 2\alpha n) = (1 - 2\alpha)n$.
- La probabilidad de que el pivote esté en una posición que produzca una partición al menos tan equilibrada es, por lo tanto:

$$
P(\text{partición equilibrada}) = \frac{(1 - 2\alpha)n}{n} = 1 - 2\alpha
$$

## Conclusión
La probabilidad de que el procedimiento PARTITION produzca una división al menos tan equilibrada como $(1-\alpha)$ a $\alpha$ es aproximadamente $1 - 2\alpha$. Esto implica que, para valores de $\alpha$ pequeños ($\alpha \leq \frac{1}{2}$), es muy probable que la división sea bastante equilibrada, lo cual es clave para la eficiencia de Quicksort en términos de complejidad esperada.


# Ejercicio 7.3-2:

RANDOMIZED-QUICKSORT selecciona un pivote al azar cada vez que particiona un subarreglo. Cada llamada al procedimiento de particionamiento implica una llamada al generador de números aleatorios RANDOM para elegir el índice del pivote.

En el peor caso: El número de llamadas al generador de números aleatorios es proporcional al número de particiones realizadas. En el peor caso, cada partición divide el subarreglo en una parte casi vacía y una parte con casi todos los elementos restantes. Por lo tanto, el número total de particiones (y el número de llamadas a RANDOM) será $Θ(n)$ para un arreglo de tamaño $n$.

En el mejor caso: En el mejor caso, cada partición divide el subarreglo en dos partes aproximadamente iguales, lo cual lleva a un árbol de particionamiento balanceado. En este caso, el número de particiones será también $Θ(n)$, ya que se realizan $n$ particiones en total, independientemente del equilibrio de las divisiones.

Por lo tanto, tanto en el mejor caso como en el peor caso, el número de llamadas al generador RANDOM es $Θ(n)$.
