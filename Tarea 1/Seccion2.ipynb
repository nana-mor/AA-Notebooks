## Ejercicio 2.1-2

Considera el procedimiento SUM-ARRAY en la página opuesta. Calcula la suma de los n números en el arreglo A[1 : n]. Explica un invariante del bucle para este procedimiento y usa sus propiedades de inicialización, mantenimiento y finalización para demostrar que el procedimiento SUM-ARRAY devuelve la suma de los números en A[1 : n].

````python
def sum_array(A, n):
    total_sum = 0
    for i in range(1, n + 1):
        total_sum += A[i]
    return total_sum
````

### Demostrando que el algoritmo SUM-ARRAY(A, n) es correcto mediante inducción matemática:

### Paso 1: Base de la inducción

Para $n = 1$ , el arreglo A[1:1] contiene un solo elemento. Según el algoritmo:

1. `sum = 0` inicialmente.
2. El bucle recorre el único valor del arreglo, sumando $A[1]$  a `sum`.
3. Después de la primera iteración, `sum = A[1]`.
4. El algoritmo devuelve `sum`.

Así, para  $n = 1$ , el procedimiento devuelve correctamente $A[1]$, lo cual es la suma de los elementos en $A[1:1]$ . La base de la inducción es verdadera.

### Paso 2: Hipótesis de inducción

Asumimos que el algoritmo funciona correctamente para  $n = k$. Es decir, suponemos que el procedimiento **SUM-ARRAY(A, k)** devuelve la suma de los primeros $k$  elementos del arreglo $A$, es decir:

$\text{SUM-ARRAY}(A, k) = A[1] + A[2] + \dots + A[k]$

### Paso 3: **Paso inductivo**

Queremos demostrar que si la hipótesis de inducción es cierta para $n = k$, entonces también es cierta para  $n = k+1$ . Es decir, queremos probar que el procedimiento **SUM-ARRAY(A, k+1)** devuelve la suma de los primeros  $k+1$  elementos del arreglo.

El algoritmo para $n = k+1$  procede de la siguiente manera:

1. Inicializa `sum = 0`.
2. En el bucle, suma los elementos  $A[1], A[2], \dots, A[k]$ , lo cual, por la hipótesis de inducción, resulta en:

$\text{sum} = A[1] + A[2] + \dots + A[k]$

1. Luego, en la siguiente iteración (la iteración $k+1$ ), el algoritmo añade $A[k+1]$ a `sum`:

$\text{sum} = A[1] + A[2] + \dots + A[k] + A[k+1]$

1. Finalmente, el algoritmo devuelve `sum`, que ahora es la suma de los primeros $k+1$ elementos del arreglo.

### Conclusión:

Hemos demostrado que:

1. El algoritmo funciona para n = 1  (base de la inducción).
2. Si el algoritmo funciona para  $n = k$ , también funciona para  $n = k+1$  (paso inductivo).

Por el principio de inducción matemática, podemos concluir que el procedimiento **SUM-ARRAY(A, n)** devuelve correctamente la suma de los primeros $n$  elementos del arreglo $A$ para cualquier  $n \geq 1$ .

Reescribe el procedimiento INSERTION-SORT para ordenar en orden decreciente en lugar de en orden creciente.

Para modificar el algoritmo de INSERTION-SORT y que ordene en orden decreciente en lugar de orden ascendente, solo necesitamos hacer un pequeño ajuste en la condición. En lugar de comparar si el elemento actual es mayor que la clave (key), debes verificar si es menor que la clave.
Así que hacemos la función donde va a ordenar una lista A en orden decreciente usando el algoritmo de Insertion Sort.Y vamos a tener: 

Parámetros:
        A (lista): La lista de elementos a ordenar.
    
Retorna:
        None: La lista A se ordena en su lugar.

def insertion_sort_descending(A):
    n = len(A)
    for i in range(1, n):  # Empezamos desde el índice 1 (corresponde a i = 2 en pseudocódigo)
        key = A[i]
        j = i - 1
        # Mueve los elementos de A[0..i-1] que son menores que la clave hacia adelante
        while j >= 0 and A[j] < key:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = key

````python
# Ejemplo de uso
if __name__ == "__main__":
    A = [5, 2, 9, 1, 5, 6]  # Lista de ejemplo
    print("Lista original:", A)
    insertion_sort_descending(A)
    print("Lista ordenada en orden decreciente:", A)
````

## Ejercicio 2.2-2

Considera ordenar n números almacenados en el arreglo A[1 : n] encontrando primero el elemento más pequeño de A[1 : n] y cambiándolo con el elemento en A[1]. Luego encuentra el elemento más pequeño de A[2 : n] y cámbialo con A[2]. Luego encuentra el elemento más pequeño de A[3 : n] y cámbialo con A[3]. Continúa de esta manera para los primeros n - 1 elementos de A. Escribe el pseudocódigo para este algoritmo, el cual es conocido como ordenamiento por selección (selection sort). ¿Qué invariante de bucle mantiene este algoritmo? ¿Por qué necesita ejecutarse solo para los primeros n - 1 elementos, en lugar de para todos los n elementos? Da el tiempo de ejecución en el peor de los casos del ordenamiento por selección en notación Θ. ¿Es mejor el tiempo de ejecución en el mejor de los casos?

### 1. Pseudocódigo para el algoritmo de Selection Sort:

El algoritmo de Selection Sort selecciona el elemento más pequeño de una lista y lo coloca en su posición correcta. Esto se repite para los primeros $n - 1$ elementos de la lista.

El pseudocódigo sería el siguiente:

```python
SelectionSort(A)
n = longitud(A)               // Obtener la longitud del arreglo
para i desde 0 hasta n-2       // Bucle para recorrer el arreglo
minIndex = i                 // Suponer que el primer elemento es el menor
para j desde i+1 hasta n-1   // Bucle para encontrar el menor elemento en el subarreglo
si A[j] < A[minIndex]      // Comparar el elemento actual con el menor encontrado
minIndex = j             // Actualizar minIndex si se encuentra un nuevo mínimo
fin para
si minIndex ≠ i              // Si minIndex ha cambiado, realizar el intercambio
intercambiar A[i] y A[minIndex] // Intercambiar el primer elemento con el menor encontrado
fin para
fin SelectionSort
```

```python
SelectionSort(array, n):
    para i desde 0 hasta n-1:
        min_index = i  // Asume que el primer elemento no ordenado es el mínimo
        
        // Encuentra el índice del elemento más pequeño en el resto del array no ordenado
        para j desde i+1 hasta n-1:
            si array[j] < array[min_index]:
                min_index = j  // Actualiza el índice del valor mínimo encontrado

        // Intercambia el elemento mínimo encontrado con el primer elemento no ordenado
        si min_index ≠ i:
            intercambiar array[i] con array[min_index]

    retornar array

```

### 2. Invariante del bucle:

Un invariante de bucle es una condición que se mantiene verdadera durante la ejecución del bucle. El invariante en este caso es que al inicio de cada iteración del bucle externo en i, los primeros $i-1$ elementos ya están ordenados y en sus posiciones correctas en el arreglo.

Esto significa que el algoritmo asegura que cada vez que selecciona un mínimo, los elementos que ya han sido procesados no necesitarán ser revisados nuevamente, ya que están en su lugar correcto.

### 3. ¿Por qué el algoritmo necesita ejecutarse solo hasta los primeros \(n - 1\) elementos?

La razón es que si colocamos correctamente los primeros $n - 1$ elementos en su lugar, el último elemento necesariamente estará en su lugar correcto. No es necesario hacer una iteración adicional sobre el último elemento, ya que será el más grande si los demás ya han sido organizados.

### 4. Tiempo de ejecución en el peor caso en notación \(\Theta\):

En el peor caso, el algoritmo de **Selection Sort** tiene que comparar cada elemento con todos los elementos restantes. Para el primer elemento, hace $n - 1$ comparaciones; para el segundo, $n - 2$; y así sucesivamente hasta la última iteración que hace 1 comparación. El número total de comparaciones es:


$(n-1) + (n-2) + \dots + 1 = \frac{n(n-1)}{2} = \frac{n^2 - n}{2}$


En términos de notación $\Theta$, el término dominante es $n^2$, por lo que el tiempo de ejecución en el peor caso es:


$\Theta(n^2)$


### 5. ¿Es mejor el tiempo de ejecución en el mejor caso?

No, el tiempo de ejecución del algoritmo de **Selection Sort** en el mejor caso también es $\Theta(n^2)$ , ya que incluso si el arreglo está ordenado, el algoritmo aún realiza las mismas comparaciones para verificar cada elemento.

### Conclusiones:

1. El pseudocódigo del algoritmo de **Selection Sort** está en la primera sección.
2. El invariante del bucle es que los primeros $i-1$  elementos ya están ordenados al inicio de cada iteración.
3. El algoritmo solo necesita ejecutarse hasta los primeros $n - 1$ elementos porque el último ya estará en su lugar correcto.
4. El tiempo de ejecución en el peor caso es  $\Theta(n^2)$ , y este también es el tiempo en el mejor caso.

El test en la línea 1 del procedimiento MERGE-SORT lee "if 
p≥r" en lugar de "if 
p
≠
r
". Si se llama a MERGE-SORT con 
p>r, entonces el subarreglo 
A[p:r] está vacío. Argumenta que, siempre y cuando la llamada inicial a MERGE-SORT(A, 1, n) tenga 
n≥1, el test "if 
p
≠
r
" es suficiente para asegurar que ninguna llamada recursiva tenga 
p>r.



### Análisis del problema:

El algoritmo de *Merge Sort* divide el arreglo recursivamente en mitades. Cada vez que se hace una llamada recursiva, se pasa un subarreglo definido por los índices $p$ y  $r$ . El propósito de la condición "if $p \geq r$ " es verificar si el subarreglo tiene uno o ningún elemento, lo cual significa que ya está ordenado.

La propuesta en el ejercicio es reemplazar esta condición por "if $p \neq r$ ". Si se cumple esta condición, entonces no llamaremos recursivamente al algoritmo cuando el subarreglo tenga un solo elemento o ningún elemento.

### Demostración de que $p \neq r$  es suficiente:

### 1. **Entendamos cuándo se cumple $p > r$** :

Cuando llamamos a *Merge Sort*, tenemos los parámetros $p$ , $r$ , y $q$ , donde:

- $p$ es el índice del primer elemento del subarreglo,
- r  es el índice del último elemento del subarreglo,
- $q$  es el índice que divide el subarreglo en dos mitades: $q = \lfloor (p + r) / 2 \rfloor$ .

Durante la recursión, se hacen dos llamadas:

- La primera llamada recursiva a la mitad izquierda: `MERGE-SORT(A, p, q)`.
- La segunda llamada recursiva a la mitad derecha: `MERGE-SORT(A, q+1, r)`.

### 2. **Cómo afecta la condición $p \neq r$** :

- Si  $p = r$ , el subarreglo contiene exactamente un elemento, por lo que ya está ordenado, y no se necesita hacer más llamadas recursivas.
- Si $p < r$ , el subarreglo tiene al menos dos elementos, y seguimos dividiendo.

Ahora, si usamos la condición "if  $p \neq r$ ", evitamos llamadas recursivas innecesarias cuando el subarreglo tiene un solo elemento (donde  $p = r$ ).

### 3. **Por qué nunca se cumple $p > r$** :

Durante la recursión, los valores de $p$  y $r$  se actualizan de tal manera que:

- En la primera llamada,  $r = q$ , por lo que siempre tenemos  $p \leq r$ .
- En la segunda llamada,  $p = q + 1$ , lo cual asegura que $p \leq r$ , ya que $q + 1$ es menor o igual que  $r$ .

Por lo tanto, la condición  $p > r$  nunca ocurre durante la ejecución de *Merge Sort*, siempre y cuando se haya iniciado correctamente con  $p = 1$  y  $r = n$ , donde  $n \geq 1$ . Como resultado, la condición "if $p \neq r$ " es suficiente para evitar llamadas innecesarias cuando el subarreglo está vacío.

### Conclusión:

La condición "if  $p \neq r$ " es suficiente para asegurar que no se haga ninguna llamada recursiva cuando  $p > r$ . En otras palabras, esta condición evita llamadas recursivas en subarreglos vacíos, lo que garantiza que el algoritmo siga funcionando correctamente sin tener que usar la condición  $p \geq r$ .

# Demostración por inducción

Para demostrar que el algoritmo `MERGE` es correcto utilizando **inducción matemática**, vamos a centrarnos en el proceso de fusión. Vamos a probar por inducción que el algoritmo `MERGE` correctamente combina los dos subarreglos \( L \) y \( R \) en un solo subarreglo ordenado \( A[p:r] \).

### Enunciado a demostrar:

Queremos demostrar que el algoritmo `MERGE(A, p, q, r)` siempre combina dos subarreglos ordenados \( L = A[p : q] \) y \( R = A[q + 1 : r] \) en un solo subarreglo ordenado A[p : r].

### Paso 1: **Caso base** (Subarreglos con un solo elemento)

Si los subarreglos L y R tienen solo un elemento cada uno, la fusión es trivial.

-  L = [A[p]]
-  R = [A[q+1]] 

Como L y R ya están ordenados por ser subarreglos con un solo elemento, el ciclo `while` de las líneas 12-18 simplemente compara \( L[0] \) con \( R[0] \), coloca el menor de los dos en \( A[p] \), y luego coloca el otro elemento en \( A[p+1] \).

- Si $ L[0] \leq R[0] $, entonces \( A[p] = L[0] \) y \( A[p+1] = R[0] \).
- Si \( R[0] < L[0] \), entonces \( A[p] = R[0] \) y \( A[p+1] = L[0] \).

En ambos casos, \( A[p : r] \) está ordenado.

**Conclusión del caso base**: Si los subarreglos tienen solo un elemento, `MERGE` funciona correctamente y ordena los dos elementos.

### Paso 2: **Hipótesis de inducción** (Suponemos que `MERGE` funciona para subarreglos de tamaño \( n \))

Supongamos que `MERGE(A, p, q, r)` funciona correctamente para dos subarreglos de tamaño \( n \). Es decir, suponemos que si \( L \) y \( R \) tienen hasta \( n \) elementos cada uno, `MERGE` correctamente los combina en un solo subarreglo ordenado \( A[p:r] \).

    - El ciclo comienza con \( i = 0 \) y \( j = 0 \), y el índice \( k = p \) para llenar el arreglo \( A \).
2. **Comportamiento del ciclo `while` (líneas 12-18)**:
    - En cada iteración, el algoritmo compara los elementos actuales de \( L \) y \( R \):
        - Si $ L[i] \leq R[j] $, entonces el menor elemento es \( L[i] \) y se copia a \( A[k] \), incrementando \( i \) y \( k \).
        - Si \( R[j] < L[i] \), entonces el menor elemento es \( R[j] \) y se copia a \( A[k] \), incrementando \( j \) y \( k \).
    
    Este proceso garantiza que después de cada iteración, el subarreglo \( A[p:k-1] \) contiene los elementos más pequeños de \( L \) y \( R \), y está ordenado.

### Conclusión del paso inductivo:

- Al asumir que `MERGE` funciona correctamente para subarreglos de tamaño \( n \), hemos demostrado que también funciona correctamente para subarreglos de tamaño \( n + 1 \).
- El proceso de comparación y fusión garantiza que todos los elementos de los subarreglos \( L \) y \( R \) se colocan en \( A \) en el orden correcto.

# **Demostración por inducción de $T(n) = n \log_2 n$**

## **Enunciado**
Queremos demostrar que, cuando $n \geq 2$ es una potencia exacta de 2, la solución de la recurrencia:

$$
T(n) = 
\begin{cases} 
2 & \text{si } n = 2, \\
2T(n/2) + n & \text{si } n > 2,
\end{cases}
$$

es $T(n) = n \log_2 n$.

---

**Paso 1: Caso base ($n = 2$)**

Para $n = 2$, la recurrencia nos da:
$$
T(2) = 2.
$$

La fórmula propuesta es $T(n) = n \log_2 n$. Sustituyendo $n = 2$ en esta fórmula:
$$
T(2) = 2 \log_2 2 = 2 \times 1 = 2.
$$

Esto coincide con la recurrencia, por lo que el **caso base** es correcto.

---

**Paso 2: Hipótesis de inducción**

Supongamos que la fórmula es correcta para algún $n = 2^k$, donde $k$ es un número entero no negativo. Es decir, suponemos que:
$$
T(2^k) = 2^k \log_2(2^k).
$$

Sabemos que $\log_2(2^k) = k$, así que:
$$
T(2^k) = 2^k \cdot k.
$$

Esta es nuestra **hipótesis de inducción**.

---

**Paso 3: Paso inductivo**

Queremos demostrar que la fórmula también es válida para $n = 2^{k+1}$. Es decir, queremos probar que:
$$
T(2^{k+1}) = 2^{k+1} \log_2(2^{k+1}).
$$

Sabemos que $\log_2(2^{k+1}) = k + 1$, así que necesitamos demostrar que:
$$
T(2^{k+1}) = 2^{k+1} \cdot (k + 1).
$$

---

**Paso 4: Sustituimos la recurrencia en $T(2^{k+1})$**

De acuerdo con la recurrencia dada:
$$
T(n) = 2T(n/2) + n.
$$

Sustituyendo $n = 2^{k+1}$, obtenemos:
$$
T(2^{k+1}) = 2T(2^k) + 2^{k+1}.
$$

---

**Paso 5: Sustituimos $T(2^k)$ usando la hipótesis de inducción**

De nuestra hipótesis de inducción, sabemos que:
$$
T(2^k) = 2^k \cdot k.
$$

Sustituyendo esto en la ecuación para $T(2^{k+1})$:
$$
T(2^{k+1}) = 2 \cdot (2^k \cdot k) + 2^{k+1}.
$$

---

**Paso 6: Simplificación paso a paso**

1. Multiplicamos los términos dentro de los paréntesis:
   $$
   T(2^{k+1}) = 2^{k+1} \cdot k + 2^{k+1}.
   $$

2. Sacamos el factor común $2^{k+1}$:
   $$
   T(2^{k+1}) = 2^{k+1} (k + 1).
   $$

---

**Paso 7: Relación con el logaritmo**

Sabemos que $\log_2(2^{k+1}) = k + 1$, así que podemos escribir:
$$
T(2^{k+1}) = 2^{k+1} \log_2(2^{k+1}).
$$

---

**Paso 8: Conclusión del paso inductivo**

Hemos demostrado que, si la fórmula es válida para $n = 2^k$, también es válida para $n = 2^{k+1}$. Esto completa el paso inductivo.

---

**Paso 9: Conclusión final**

Por **inducción matemática**, hemos demostrado que la solución de la recurrencia es:
$$
T(n) = n \log_2 n,
$$
para cualquier $n \geq 2$ que sea una potencia exacta de 2.

- **Caso base**: Verificamos que la fórmula es correcta para $n = 2$.
- **Paso inductivo**: Demostramos que si la fórmula es válida para $n = 2^k$, entonces también es válida para $n = 2^{k+1}$.

## Ejercicio 2.3-6

### Enunciado:

Piensa en el algoritmo de *Insertion Sort* como un algoritmo recursivo. Para ordenar el arreglo A[1:n], primero ordena recursivamente el subarreglo A[1:n-1] y luego inserta A[n] en el subarreglo ordenado A[1:n-1]. Escribe el pseudocódigo para esta versión recursiva de *Insertion Sort*.

### Paso 1: **Idea del algoritmo**

El algoritmo de *Insertion Sort* normalmente funciona insertando cada elemento en su posición correcta dentro del subarreglo que ya está ordenado. En esta versión recursiva:

1. **Ordenamos recursivamente el subarreglo A[1:n-1]**.
2. **Insertamos A[n] en su lugar correcto dentro del subarreglo ya ordenado A[1:n-1]**.

Este proceso se repite hasta que el subarreglo a ordenar tenga solo un elemento (un arreglo con un solo elemento ya está ordenado).

### Paso 2: **Pseudocódigo**

Aquí está el pseudocódigo para la versión recursiva de *Insertion Sort*:

INSERTION-SORT-RECURSIVE(A, n)
1. if n <= 1
2.     return
3. INSERTION-SORT-RECURSIVE(A, n-1)
4. key = A[n]
5. i = n - 1
6. while i > 0 and A[i] > key
7.     A[i+1] = A[i]
8.     i = i - 1
9. A[i+1] = key

### Explicación del pseudocódigo:

1. **Caso base**: Si n ≤ 1, el subarreglo tiene uno o ningún elemento y ya está ordenado, por lo que no se hace nada (líneas 1-2).
2. **Llamada recursiva**: Ordenamos recursivamente el subarreglo A[1:n-1] (línea 3). Esto reduce el problema al tamaño n-1.
3. **Inserción**: Después de que el subarreglo A[1:n-1] está ordenado, almacenamos el valor A[n] en la variable `key` y comenzamos a mover los elementos de A hacia la derecha para insertar el valor en su posición correcta (líneas 4-9).
   - **Línea 6**: Mientras el valor de A[i] sea mayor que la `key`, desplazamos los elementos hacia la derecha.
   - **Línea 9**: Cuando encontramos la posición correcta, insertamos `key` en su lugar.

### Paso 3: **Recurrencia del tiempo de ejecución (peor caso)**

Para analizar el tiempo de ejecución, necesitamos establecer la recurrencia que describe el costo del algoritmo.

- En cada llamada recursiva, se hace una llamada a INSERTION-SORT-RECURSIVE(A, n-1) (línea 3), y luego se inserta el valor A[n] en su posición correcta.
- El tiempo para insertar A[n] es lineal en el peor caso (es decir, podría tener que comparar con todos los elementos de A[1:n-1]).

Por lo tanto, el tiempo total T(n) del algoritmo en el peor caso se describe con la siguiente **recurrencia**:

T(n) = T(n-1) + O(n)

### Solución de la recurrencia:

Esta recurrencia es similar a la de la versión iterativa de *Insertion Sort*. Vamos a resolverla paso a paso:

1. Expandimos la recurrencia:

   T(n) = T(n-1) + n

   T(n-1) = T(n-2) + (n-1)

   T(n-2) = T(n-3) + (n-2)

   ...

   T(2) = T(1) + 2

2. Sumamos todas las ecuaciones:

   T(n) = T(1) + 2 + 3 + ... + n

3. Sabemos que T(1) = O(1), ya que un subarreglo de un solo elemento ya está ordenado.

4. La suma 2 + 3 + ... + n es una **progresión aritmética**, cuya fórmula es:

   sum(k=2 to n) k = (n(n+1)/2) - 1 = O(n^2)

Por lo tanto, el tiempo de ejecución del algoritmo es T(n) = O(n^2) en el peor caso, que es el mismo tiempo de ejecución que la versión iterativa de *Insertion Sort*.

### Resumen:

- Hemos escrito el pseudocódigo para una versión recursiva de *Insertion Sort*.
- La recurrencia que describe el tiempo de ejecución es T(n) = T(n-1) + O(n), que se resuelve a O(n^2) en el peor caso, igual que la versión iterativa.


## Ejercicio 2.3-7:

**Enunciado:**
El bucle `while` de las líneas 5-7 del procedimiento *Insertion Sort* en la Sección 2.1 utiliza una búsqueda lineal para recorrer hacia atrás el subarreglo ordenado $A[1:j-1]$. ¿Qué sucedería si el algoritmo *Insertion Sort* utilizara una búsqueda binaria en lugar de una búsqueda lineal para encontrar la posición correcta para insertar $A[j]$? ¿Mejoraría esto el tiempo de ejecución en el peor caso de *Insertion Sort* a $\Theta(n \log n)$?

Paso 1: **Recordemos cómo funciona el Insertion Sort original**

En el algoritmo *Insertion Sort* original, en cada paso $j$, buscamos la posición correcta para insertar el valor $A[j]$ dentro del subarreglo ya ordenado $A[1:j-1]$. Esto se hace utilizando una **búsqueda lineal** (comparando el valor de $A[j]$ con cada elemento de $A[1:j-1]$ hasta encontrar su posición correcta). Luego, se desplazan los elementos mayores a la derecha para hacer espacio para $A[j]$.

El pseudocódigo para la búsqueda lineal dentro de *Insertion Sort* es más o menos así:

for j = 2 to n
    key = A[j]
    i = j - 1
    while i > 0 and A[i] > key
        A[i+1] = A[i]
        i = i - 1
    A[i+1] = key

Paso 2: **Cómo funcionaría si usamos búsqueda binaria**

En lugar de usar una **búsqueda lineal** para encontrar la posición correcta de $A[j]$, podemos utilizar una **búsqueda binaria** porque sabemos que el subarreglo $A[1:j-1]$ ya está ordenado. La búsqueda binaria permite encontrar la posición correcta en $O(\log j)$ tiempo.

La idea es:

1. Usar **búsqueda binaria** en el subarreglo ordenado $A[1:j-1]$ para encontrar la posición donde debería insertarse $A[j]$.
2. Desplazar todos los elementos mayores que $A[j]$ hacia la derecha, lo que sigue tomando tiempo $O(j)$.

Pseudocódigo de Insertion Sort con Búsqueda Binaria

El pseudocódigo para el *Insertion Sort* usando **búsqueda binaria** para encontrar la posición correcta sería algo así:

for j = 2 to n
    key = A[j]
    pos = BINARY-SEARCH(A, key, 1, j-1)  // Buscar la posición correcta
    i = j - 1
    while i ≥ pos
        A[i+1] = A[i]  // Desplazar los elementos hacia la derecha
        i = i - 1
    A[pos] = key  // Insertar el valor en la posición correcta

Aquí, la función `BINARY-SEARCH(A, key, 1, j-1)` es una búsqueda binaria que devuelve el índice donde debería insertarse el valor `key`.

Paso 3: **Recurrencia del tiempo de ejecución con búsqueda binaria**

Ahora vamos a analizar el tiempo de ejecución de este algoritmo modificado. El tiempo de ejecución total es la suma de:

1. **Tiempo de búsqueda binaria**: En cada paso $j$, la búsqueda binaria se realiza sobre el subarreglo $A[1:j-1]$, lo que toma $O(\log j)$ tiempo.
2. **Tiempo de desplazamiento de elementos**: Una vez que encontramos la posición correcta con la búsqueda binaria, necesitamos desplazar los elementos mayores a la derecha. Esto sigue tomando $O(j)$ tiempo en el peor caso, porque podemos tener que desplazar hasta $j - 1$ elementos.

El tiempo total para el paso $j$ es la suma del tiempo de búsqueda binaria y el tiempo de desplazamiento de elementos:

$T_j = O(\log j) + O(j)$.

Paso 4: **Tiempo total de ejecución**

Para calcular el tiempo total del algoritmo, sumamos el tiempo para cada paso $j$ desde $2$ hasta $n$:

$T(n) = \sum_{j=2}^n (O(\log j) + O(j))$.

Dividimos la suma en dos partes:

1. **Suma del tiempo de búsqueda binaria**:

$\sum_{j=2}^n O(\log j) = O(\log 2 + \log 3 + \dots + \log n)$.

   Utilizando la propiedad de los logaritmos $\log a + \log b = \log(ab)$, podemos simplificar esta suma:

$\sum_{j=2}^n O(\log j) = O(\log(n!))$.

   Usando la **aproximación de Stirling** para $n!$, sabemos que $\log(n!) = O(n \log n)$. Por lo tanto:

$\sum_{j=2}^n O(\log j) = O(n \log n)$.

2. **Suma del tiempo de desplazamiento**:

$\sum_{j=2}^n O(j) = O(2 + 3 + \dots + n)$.

   Esta es una **progresión aritmética**, y la suma de los primeros $n-1$ términos de una progresión aritmética es:

$\sum_{j=2}^n O(j) = O(n^2)$.

Paso 5: **Conclusión sobre el tiempo de ejecución**

El tiempo total del algoritmo es la suma de los dos términos:

$T(n) = O(n \log n) + O(n^2)$.

El término dominante es $O(n^2)$, lo que significa que el tiempo de ejecución **no mejora** el tiempo en el peor caso. Incluso usando búsqueda binaria, el tiempo de desplazamiento de los elementos sigue tomando $O(n^2)$ en total, lo que hace que el tiempo de ejecución final sea $O(n^2)$.

Respuesta:

Aunque usamos búsqueda binaria para encontrar la posición correcta, el tiempo de ejecución en el peor caso sigue siendo $O(n^2)$ debido al tiempo que toma desplazar los elementos en el arreglo. **No mejora** el tiempo de ejecución a $\Theta(n \log n)$.

## Ejercicio 2.3-8:

**Enunciado:**
Describe un algoritmo que, dado un conjunto $ S $ de $ n $ enteros y otro entero $ x $, determine si $ S $ contiene dos elementos que suman $ x $. Tu algoritmo debe ejecutarse en el tiempo $ \Theta(n \log n) $ en el peor caso.

Paso 1: **Idea básica del problema**

El problema nos pide verificar si en el conjunto S de n enteros existen dos elementos a y b tales que a + b = x . Debemos hacerlo con un tiempo de ejecución $\Theta(n \log n)$, lo que indica que debemos usar un enfoque más eficiente que una búsqueda doble anidada, que tendría un tiempo $ O(n^2) $.

La clave está en **ordenar el conjunto** y luego usar un enfoque eficiente de búsqueda, como el uso de dos punteros o búsqueda binaria. Ordenar un conjunto toma $ O(n \log n) $, y luego usar búsqueda binaria o un enfoque de punteros puede hacer que el problema se resuelva en tiempo lineal $ O(n) $.

Paso 2: **Algoritmo usando el enfoque de dos punteros**

Aquí está el plan del algoritmo:

1. Primero, **ordenamos** el conjunto $ S $.
2. Luego, usamos dos punteros, uno al inicio y otro al final del conjunto, para encontrar dos números que sumen $ x $.
    - Si la suma de los dos números es igual a $ x $, hemos encontrado la solución.
    - Si la suma es menor que $ x $, movemos el puntero izquierdo hacia la derecha para aumentar la suma.
    - Si la suma es mayor que $ x $, movemos el puntero derecho hacia la izquierda para reducir la suma.

Paso 3: **Pseudocódigo**

```python
# TWO-SUM(S, x)
def two_sum(S, x):
    # 1. Ordenar S de forma ascendente  // Esto toma O(n log n)
    S.sort()
    
    # 2. Inicializar dos punteros: izq = 0 (primera posición), der = len(S) - 1 (ultima posición)
    izq = 0
    der = len(S) - 1
    
    # 3. Mientras izq < der hacer:
    while izq < der:
        # 4. sum = S[izq] + S[der]
        suma = S[izq] + S[der]
        
        # 5. Si sum == x
        if suma == x:
            return True  # Hemos encontrado dos números que suman x
        
        # 7. Si sum < x
        elif suma < x:
            izq += 1  # Mover el puntero izquierdo a la derecha para aumentar la suma
        
        # 9. Si sum > x
        else:
            der -= 1  # Mover el puntero derecho a la izquierda para reducir la suma
    
    # 11. Retornar FALSE  // No hay dos números que sumen x
    return False
```

Explicación del código:

1. **Ordenar el conjunto**: El primer paso es ordenar el conjunto $ S $. Esto toma $ O(n \log n) $, que es el tiempo requerido para la ordenación.
2. **Dos punteros**:
    - Inicializamos el puntero izquierdo $ izq $ en la primera posición del conjunto, es decir, $ izq = 0 $.
    - Inicializamos el puntero derecho $ der $ en la última posición del conjunto, es decir, $ der = n - 1 $.
3. **Bucle principal**:
    - Mientras $ izq $ sea menor que $ der $, calculamos la suma de los valores en $ S[izq] $ y $ S[der] $.
    - Si la suma es igual a $ x $, retornamos `True`, lo que indica que hemos encontrado dos números que suman $ x $.
    - Si la suma es menor que $ x $, movemos el puntero izquierdo hacia la derecha, porque necesitamos una suma más grande.
    - Si la suma es mayor que $ x $, movemos el puntero derecho hacia la izquierda, porque necesitamos una suma más pequeña.
4. **Si no encontramos dos números** que sumen $ x $ después de que los punteros se crucen, retornamos `False`, lo que significa que no existen dos números en $ S $ que sumen $ x $.

 Paso 4: **Análisis del tiempo de ejecución**

El tiempo de ejecución del algoritmo es:

1. **Ordenar el conjunto $ S $**: Ordenar un conjunto de $ n $ elementos toma $ O(n \log n) $.
2. **Bucle con dos punteros**: Después de ordenar el conjunto, recorremos el conjunto una vez con dos punteros, lo que toma tiempo lineal $ O(n) $.

Por lo tanto, el tiempo total del algoritmo es:

$$
T(n) = O(n \log n) + O(n) = O(n \log n).
$$

Este tiempo cumple con la condición de ejecutar el algoritmo en $ \Theta(n \log n) $. 

Paso 5: **Ejemplo paso a paso**

Supongamos que tenemos el conjunto $ S = [1, 4, 5, 6, 8] $ y queremos verificar si hay dos números que sumen $ x = 9 $.

1. **Ordenamos $ S $** (ya está ordenado en este caso):
$S = [1, 4, 5, 6, 8]$
2. **Inicializamos dos punteros**:
    - $ izq = 0 $ (primer elemento, valor 1).
    - $ der = 4 $ (ultimo elemento, valor 8).
3. **Primera iteración**:
    - Calculamos la suma $ suma = S[0] + S[4] = 1 + 8 = 9 $.
    - Como $ suma = 9 $, retornamos `True`, ya que encontramos dos números que suman $ x = 9 $.

Para el conjunto $ S = [2, 3, 7, 11, 15] $ y $ x = 10 $:

1. **Ordenamos $ S $** (ya está ordenado):
$S = [2, 3, 7, 11, 15]$
2. **Inicializamos dos punteros**:
    - $ izq = 0 $, valor 2.
    - $ der = 4 $, valor 15.
3. **Primera iteración**:
