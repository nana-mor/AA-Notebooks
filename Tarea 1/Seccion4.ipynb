## Enunciado:
Utilizando el método de sustitución resuelve:

$$
T(n) = T(n - 1) + n
$$

suponiendo que tiene una solución $T(n) = O(n^2)$.

### Hipótesis inductiva

Suponemos que la solución para $T(n)$ es $O(n^2)$. Suponemos que existe una constante $c$ tal que:

$$
T(n) \leq c \cdot n^2
$$

### Caso base 

Tomamos como $T(1)$. Suponemos:

$$
T(1) = 1
$$

Entonces, $c$ debe ser lo suficientemente grande para que:

$$
T(1) = 1 \leq c \cdot 1^2 = c
$$

### Hipótesis inductiva 

Suponiendo que $T(k) \leq c \cdot k^2$ es cierto para todos los $k < n$, aplicamos la sustitución para $T(n-1)$:

$$
T(n) = T(n - 1) + n
$$
$$
T(n) \leq c \cdot (n - 1)^2 + n
$$


Expresamos $c \cdot (n - 1)^2$ en términos de $n$:

$$
c \cdot (n - 1)^2 = c \cdot (n^2 - 2n + 1)
$$
$$
T(n) \leq c \cdot n^2 - 2cn + c + n
$$

Para que $T(n) \leq c \cdot n^2$, necesitamos:

$$
-2cn + c + n \leq 0
$$
$$
n(1 - 2c) + c \leq 0
$$

Podemos ajustar $c$ para que esta desigualdad se mantenga, eligiendo $c$ suficientemente grande para que el término lineal $n(1 - 2c)$ sea dominado por $-2cn$, manteniendo $c$ mayor o igual que el término libre. Por ejemplo:

$$
c = 1
$$


Al probarlo con $c = 1$, notamos que:

$$
T(n) \leq n^2 - 2n + 1 + n = n^2 - n + 1
$$

Esto no cumple directamente $T(n) \leq n^2$ para todo $n$, así que necesitamos ajustar $c$. Escogiendo $c$ suficientemente grande, por ejemplo $c = 1$, y luego ajustando para que el término lineal y constante sean absorbidos por $c \cdot n^2$, nos aseguramos que la desigualdad se cumple.

Así podemos afirmar que $T(n) = O(n^2)$ es una cota correcta para esta recurrencia usando el método de sustitución, eligiendo $c$ adecuadamente grande para cubrir todos los casos. Esta solución asume que $c$ puede ser cualquier constante que cumpla las condiciones.

demostrar que tiene solución $T(n) = O(\log n)$.

### Hipótesis inductiva
Suponemos que $T(n) = O(\log n)$. Esto implica que existen constantes $c > 0$ y $n_0 \geq 1$ tales que para todo $n \geq n_0$,

$$
T(n) \leq c \cdot \log n
$$

### Base inductiva
Supongamos que $T(1) = 0$. Este caso base es trivialmente menor o igual a $c \cdot \log 1 = 0$, para cualquier $c \geq 0$.

### Paso inductivo
Suponemos que la afirmación es cierta para $n/2$, es decir, $T(n/2) \leq c \cdot \log(n/2)$. Sustituimos esto en la recurrencia para obtener $T(n)$:

$$
T(n) = T(n/2) + \Theta(1)
$$
$$
\leq c \cdot \log(n/2) + \Theta(1)
$$

Por la propiedad de los logaritmos $ \log(n/2) = \log n - \log 2 $. Como $ \log 2 = 1 $, esto se convierte en:

$$
T(n) \leq c \cdot (\log n - 1) + \Theta(1)
$$
$$
= c \cdot \log n - c + \Theta(1)
$$

### Paso 5: Ajuste de la constante $c$
Para asegurar que $T(n) \leq c \cdot \log n$ es válido para todos los $n \geq n_0$, necesitamos que $-c + \Theta(1)$ sea absorbido por el término $c \cdot \log n$. Esto se puede garantizar eligiendo $c$ y la constante implícita en $\Theta(1)$ adecuadamente. Supongamos que $\Theta(1)$ es una constante positiva $k$, entonces:

$$
-c + k \leq 0
$$
$$
c \geq k
$$

Eligiendo $c$ al menos tan grande como $k$ y considerando $n_0 \geq 1$, aseguramos que la inecuación se mantiene.

Al ajustar $c$ adecuadamente y partiendo de la base de que $\log n$ aumenta muy lentamente, cualquier término constante $k$ puede ser cubierto por $c \cdot \log n$. Esto muestra que $T(n) = O(\log n)$ es una cota correcta para esta recurrencia, demostrando que crece en orden logarítmico respecto a $n$.

Vamos a suponer que $T(n) = c \cdot n \log n$ para algún $c > 0$. Necesitamos probar que esto satisface la recurrencia para todos los $n \geq 2$, asumiendo que es cierto para $n/2$.

Suponemos que la propiedad es cierta para $n/2$, es decir:

$$
T(n/2) = c \cdot \frac{n}{2} \log \frac{n}{2}
$$

Sustituimos la hipótesis inductiva en la recurrencia:

$$
T(n) = 2 \left(c \cdot \frac{n}{2} \log \frac{n}{2}\right) + n
$$

Utilizando la propiedad del logaritmo $ \log \frac{n}{2} = \log n - \log 2 $, donde $ \log 2 = 1 $:

$$
T(n) = 2 \left(c \cdot \frac{n}{2} (\log n - 1)\right) + n
$$
$$
T(n) = c \cdot n \log n - c \cdot n + n
$$
$$
T(n) = c \cdot n \log n - (c - 1) \cdot n
$$

Queremos que $T(n)$ siga siendo $c \cdot n \log n$. Para que esto ocurra, el término $- (c - 1) \cdot n$ debe ser absorbido por el término dominante $c \cdot n \log n$, podemos elegir $c$ de manera que esto se cumpla. Una opción es simplemente asegurarse de que $c$ sea suficientemente grande para que el término constante adicional no afecte la forma asintótica.

Si $c$ es suficientemente grande, entonces $T(n) \leq c \cdot n \log n$ se cumple, lo que implica que $T(n) = O(n \log n)$. Además, debido a la forma de la recurrencia, donde cada división se suma $n$, es fácil ver que $T(n)$ no puede ser menor que $O(n \log n)$, estableciendo así que $T(n) = \Theta(n \log n)$.

Esto concluye la demostración de que $T(n)$ tiene una complejidad de $ \Theta(n \log n) $.

$$
T(n) \leq c \cdot n \log n

Suponemos que la afirmación es cierta para $n/2 + 17$, es decir, 

$$
T(n/2 + 17) \leq c \cdot \left(\frac{n}{2} + 17\right) \log\left(\frac{n}{2} + 17\right).
$$ 

Sustituimos esto en la recurrencia para obtener $T(n)$:

$$
T(n) = 2T(n/2 + 17) + n
$$
$$
\leq 2\left[c \cdot \left(\frac{n}{2} + 17\right) \log\left(\frac{n}{2} + 17\right)\right] + n
$$

Utilizamos la propiedad de los logaritmos y la naturaleza del logaritmo para manejar el término $n/2 + 17$. Notamos que $\log(n/2 + 17)$ es aproximadamente $\log n$ para grandes $n$, dado que 17 es constante y se vuelve menos significativo a medida que $n$ aumenta:

$$
T(n) \leq 2\left[c \cdot \left(\frac{n}{2} + 17\right) \log n\right] + n
$$
$$
= c \cdot n \log n + 34c \cdot \log n + n
$$

Para asegurar que $T(n) \leq c \cdot n \log n$ es válido para todos los $n \geq n_0$, necesitamos que el término adicional $34c \cdot \log n + n$ sea absorbido por el término $c \cdot n \log n$. Esto puede ser garantizado si el $c$ elegido es suficientemente grande para que el crecimiento adicional sea menor en comparación con el término principal:

$$
c \cdot n \log n \geq 34c \cdot \log n + n
$$
$$
n \log n \geq 34 \log n + \frac{n}{c}
$$

Por tanto, $c$ debe ser al menos tan grande como para que el término $\frac{n}{c}$ sea menor que $n \log n$ para grandes $n$. Si ajustamos $c$ para manejar estos términos adicionales, podemos concluir que $T(n) = O(n \log n)$.


Este ajuste muestra que $T(n) = O(n \log n)$ es una cota correcta para esta recurrencia, demostrando que el crecimiento sigue siendo $O(n \log n)$ a pesar de la adición de la constante en el término recursivo.

$$
c_1 \cdot n \leq T(n) \leq c_2 \cdot n

$$
c_1 \cdot \frac{n}{3} \leq T(n/3) \leq c_2 \cdot \frac{n}{3}
$$


$$
T(n) = 2T(n/3) + \Theta(n)
$$

$$
T(n) \geq 2\left(c_1 \cdot \frac{n}{3}\right) + kn
$$
$$
T(n) \leq 2\left(c_2 \cdot \frac{n}{3}\right) + kn
$$

Esto se simplifica a:

$$
T(n) \geq \frac{2c_1n}{3} + kn

$$
T(n) \leq \frac{2c_2n}{3} + kn
$$

$$
\frac{2c_1}{3}n + kn \leq cn \leq \frac{2c_2}{3}n + kn
$$

Esto implica que podemos elegir $c_1, c_2$ de tal manera que absorban el término $kn$ y que aún se mantenga el comportamiento lineal de $T(n)$. Por ejemplo, podríamos establecer:

c_1 = \frac{k}{3}

c_2 = k

La elección de las constantes adecuadas para $c_1$ y $c_2$, y una constante $k$ que representa $\Theta(n)$, nos permitirá concluir que $T(n) = \Theta(n)$, demostrando que la recurrencia crece linealmente con $n$.

$$
c_1 \cdot n^2 \leq T(n) \leq c_2 \cdot n^2

### Paso 3: Paso inductivo

$$
c_1 \cdot \left(\frac{n}{2}\right)^2 \leq T(n/2) \leq c_2 \cdot \left(\frac{n}{2}\right)^2

T(n) = 4T(n/2) + \Theta(n)

## Ejercicio 4.3-2

El ejercicio 4.3-2 se centra en la recurrencia $T(n) = 4T(n/2) + n$ y nos pide mostrar un intento de demostración usando el método de sustitución con la suposición $T(n) \leq c n^2$ y explicar por qué puede fallar. Luego, debemos mostrar cómo hacer que un método de sustitución funcione, probablemente ajustando la suposición.

### Intento fallido con $T(n) \leq c n^2$

**Paso 1: Hipótesis inductiva**

Suponemos que $T(n) \leq c n^2$ para un $c$ adecuado y para todo $n \geq n_0$.

**Paso 2: Sustitución en la recurrencia**

$T(n) = 4T(n/2) + n$

Asumiendo que la hipótesis inductiva es verdadera para $n/2$:

$T(n) \leq 4(c(n/2)^2) + n$

$T(n) \leq c n^2 + n$

**Paso 3: Análisis de la expresión**

La expresión resultante $c n^2 + n$ claramente no está acotada por $c n^2$ para todos los valores de $n$ debido al término lineal $n$, que no puede ser absorbido por el término cuadrático para valores grandes de $n$ a menos que $c$ sea ajustado de manera que $n$ sea insignificante en comparación, lo cual es imposible en términos asintóticos.

### Método correcto para hacer que la sustitución funcione

**Nuevo enfoque**

Necesitamos ajustar la hipótesis inductiva para incluir un término que maneje correctamente el término lineal adicional $n$. Podemos intentar agregar un término lineal a nuestra suposición:

$T(n) \leq c n^2 + d n$

donde $d$ es otra constante que necesitamos determinar.

**Paso 1: Hipótesis inductiva ajustada**

Suponemos que $T(n) \leq c n^2 + d n$ para todos los $n \geq n_0$.

**Paso 2: Sustitución en la recurrencia**

$T(n) = 4T(n/2) + n$

$T(n) \leq 4(c(n/2)^2 + d(n/2)) + n$

$T(n) \leq c n^2 + 2 d n + n$

**Paso 3: Ajuste de las constantes**

Queremos que $c n^2 + 2 d n + n$ esté acotado por $c n^2 + d n$. Esto se cumple si:

$2 d n + n \leq d n$

$n \leq d n$

Esto requiere que $d$ sea al menos 1. Si ajustamos $d$ para ser mayor o igual a 1, entonces $2 d n + n$ siempre será menor o igual a $d n$ para $n \geq 1$ si $d \geq 1$.

### Conclusión

Este ajuste muestra que si consideramos una forma más compleja para la hipótesis inductiva que incluya términos adicionales para manejar los términos residuales, podemos hacer que la demostración de la sustitución funcione. Ajustar $d$ adecuadamente (al menos igual a 1) garantiza que la recurrencia está correctamente acotada por $c n^2 + d n$, lo que demuestra que $T(n) = O(n^2)$ es una cota correcta para esta recurrencia.


# Ejercicio 4.3-3

Para resolver el ejercicio 4.3-3, abordaremos la recurrencia

$T(n) = 2T(n-1) + 1$

y nos piden demostrar que un intento de prueba utilizando el método de sustitución con la suposición $T(n) \leq c \cdot 2^n$ falla. Luego, exploraremos cómo ajustar el método de sustitución para hacer que la prueba funcione.

### Intento de prueba fallido con $T(n) \leq c \cdot 2^n$

#### Paso 1: Hipótesis inductiva

Supongamos que $T(n) \leq c \cdot 2^n$ para algún $c > 0$ y para todo $n \geq n_0$.

#### Paso 2: Sustitución en la recurrencia

Si la hipótesis inductiva es verdadera para $n-1$, entonces: $T(n-1) \leq c \cdot 2^{n-1}$

Sustituyendo esto en la recurrencia:

$T(n) = 2T(n-1) + 1$

$T(n) \leq 2(c \cdot 2^{n-1}) + 1$

$T(n) \leq c \cdot 2^n + 1$

#### Paso 3: Análisis de la expresión

Observamos que la expresión $c \cdot 2^n + 1$ no está necesariamente acotada por $c \cdot 2^n$ para todos los valores de $n$, ya que el término constante $+1$ se vuelve problemático. No podemos simplemente absorber $+1$ dentro del término dominante $c \cdot 2^n$ a medida que $n$ crece, porque siempre será adicional y afectará la igualdad, especialmente para valores pequeños de $n$.

### Método correcto para hacer que la sustitución funcione

#### Nuevo enfoque

Para ajustar el método de sustitución y permitir que la demostración funcione, consideramos una modificación en la forma de la función propuesta para $T(n)$.

#### Paso 1: Hipótesis inductiva ajustada

Proponemos una nueva hipótesis inductiva, considerando un ajuste en el término constante:

$T(n) \leq c \cdot 2^n - 1$

#### Paso 2: Sustitución en la recurrencia

Si $T(n-1) \leq c \cdot 2^{n-1} - 1$:

$T(n) = 2T(n-1) + 1$

$T(n) \leq 2(c \cdot 2^{n-1} - 1) + 1$

$T(n) \leq c \cdot 2^n - 2 + 1$

$T(n) \leq c \cdot 2^n - 1$

#### Paso 3: Verificación y ajuste de la constante $c$

El ajuste $-1$ en la hipótesis inductiva garantiza que podemos manejar el $+1$ en la recurrencia. Este término $-1$ se duplica y luego se anula parcialmente con el $+1$ adicional, lo que permite mantener la forma ajustada y demuestra que $T(n) = O(2^n)$.

### Conclusión

Al ajustar la forma de la hipótesis inductiva para incluir un término que compensa consistentemente el término constante adicional en la recurrencia, podemos hacer que la demostración de la sustitución funcione correctamente. La nueva forma de $T(n) \leq c \cdot 2^n - 1$ asegura que la recurrencia está correctamente acotada, y demostramos que $T(n) = O(2^n)$, manteniendo la estructura y el comportamiento exponencial de la recurrencia.


# Ejercicio 4.4-1

Para resolver el ejercicio 4.4-1 a) de la recurrencia $T(n) = T(n/2) + n^3$, vamos a esbozar primero su árbol de recursión, conjeturar una cota superior asintótica, y luego usar el método de sustitución para verificar la conjetura.

### Esbozo del árbol de recursión
En el árbol de recursión para $T(n)$:

- En el nivel 0, tenemos $T(n)$.
- En el nivel 1, el costo se divide en $T(n/2)$ más el costo de $n^3$ para la operación en el nivel.
- En el nivel 2, cada nodo tiene un costo $T(n/4)$ y $(n/2)^3$.
- Y así sucesivamente.

Para cada nivel $i$, el tamaño del subproblema es $n/2^i$ y el costo adicional en cada nivel es $(n/2^i)^3$.

### Cálculo del costo total del árbol de recursión
El número de niveles del árbol es aproximadamente $\log_2 n$ porque en cada nivel el tamaño del problema se reduce a la mitad. El costo en el nivel $i$ es $(n/2^i)^3$, y el número de nodos en cada nivel es $2^i$. Por lo tanto, el costo total del árbol se puede calcular como:

$$
\sum_{i=0}^{\log_2 n} 2^i \left(\frac{n}{2^i}\right)^3 = \sum_{i=0}^{\log_2 n} \frac{n^3}{2^{2i}} = n^3 \sum_{i=0}^{\log_2 n} \frac{1}{4^i}
$$

La serie geométrica $\sum_{i=0}^{\log_2 n} \frac{1}{4^i}$ converge a $\frac{4}{3}$ para $\log_2 n$ términos cuando $n$ tiende a infinito, lo que hace que el costo total sea aproximadamente $\frac{4}{3} n^3$.

### Conjetura
Por lo tanto, la conjetura para la cota superior asintótica es $T(n) = O(n^3)$.

### Verificación por el método de sustitución
Para verificar, asumimos que $T(n) \leq c n^3$ para algún $c > 0$ y todos los $n$ adecuados, y mostramos que esto se mantiene para $n$:

$$
T(n) = T(n/2) + n^3 \leq c \left(\frac{n}{2}\right)^3 + n^3 = \frac{c}{8} n^3 + n^3
$$

Queremos que $\frac{c}{8} n^3 + n^3 \leq c n^3$. Simplificando, encontramos:

$$
\frac{c}{8} + 1 \leq c \implies \frac{7c}{8} \geq 1 \implies c \geq \frac{8}{7}
$$

Elegimos $c$ al menos $\frac{8}{7}$ para satisfacer la inecuación, verificando que $T(n) = O(n^3)$ es una cota válida. Esta es una elección válida para probar que la conjetura es correcta, y $T(n) = O(n^3)$.

## Ejercicio 4.4-1 b)
Para resolver la recurrencia $T(n) = 4T(n/3) + n$, sigamos el mismo enfoque: esbozar el árbol de recursión, calcular el costo total, conjeturar una cota superior asintótica y verificar la conjetura usando el método de sustitución.

### Esbozo del árbol de recursión
En el nivel 0, el costo es $n$ (el término $n$).  
En el nivel 1, hay 4 subproblemas de tamaño $n/3$, cada uno con un costo adicional de $n/3$, resultando en un costo total de $4 \cdot T(n/3) + n$.  
En el nivel 2, cada subproblema genera 4 subproblemas más de tamaño $n/9$, y el costo en ese nivel es $4^2 \cdot T(n/9) + n/3 + n$.  
Generalizando:  
En el nivel $i$, hay $4^i$ subproblemas, cada uno de tamaño $n/3^i$.  
El costo adicional en el nivel $i$ es $4^i \cdot (n/3^i) = n \cdot (4/3)^i$.

### Cálculo del costo total
El costo total del árbol de recursión es la suma de los costos de todos los niveles:

$$
T(n) = \sum_{i=0}^{\log_3 n} n \cdot \left( \frac{4}{3} \right)^i
$$

Esto es una serie geométrica con razón $\frac{4}{3}$. La suma de una serie geométrica con $k$ términos está acotada por:

$$
\text{Suma} = a \cdot \frac{1 - r^k}{1 - r}, \quad \text{donde } a = n, \, r = \frac{4}{3}
$$

Para $k = \log_3 n$, la suma domina el término inicial y crece como $O(n^{\log_3 4})$, ya que $\log_3 4$ es la tasa de crecimiento de los nodos totales en el árbol.

### Conjetura
La cota asintótica superior para $T(n)$ es $T(n) = O(n^{\log_3 4})$. Dado que $\log_3 4 \approx 1.26$, tenemos que $T(n) = O(n^{1.26})$.

### Verificación por el método de sustitución
Hipótesis inductiva: Supongamos que $T(n) \leq c \cdot n^{\log_3 4}$ para algún $c > 0$.  
Sustituyamos en la recurrencia:

$$
T(n) = 4T(n/3) + n
$$

Sustituimos la hipótesis inductiva:

$$
T(n) \leq 4 \cdot c \left( \frac{n}{3} \right)^{\log_3 4} + n
$$

Simplifiquemos:

$$
T(n) \leq 4 \cdot c \cdot n^{\log_3 4} \cdot 3^{-\log_3 4} + n
$$

Considera la función $f(n) = \lg n$. Argumenta que, aunque $f(n/2) < f(n)$, la condición de regularidad $a f(n/b) \leq c f(n)$ con $a = 1$ y $b = 2$ no se cumple para ninguna constante $c < 1$. Además, argumenta que, para cualquier $\epsilon > 0$, la condición en el caso 3 que $f(n) = \Omega(n^{\log_b a + \epsilon})$ no se cumple.

# Solución

La función dada es $f(n) = \lg n$, y estamos trabajando con $a = 1$ y $b = 2$.

## 1. Verificación de la Condición de Regularidad

La condición de regularidad es:

$$a f(n/b) \leq c f(n) \quad \text{para alguna constante } c < 1$$

Sustituyendo los valores de $a$, $b$, y $f(n)$:

- $a = 1$
- $b = 2$
- $f(n) = \lg n$

Entonces:

$$f(n/b) = \lg(n/2) = \lg n - \lg 2 = \lg n - 1$$

La condición de regularidad se convierte en:

$$f(n/2) = \lg n - 1 \leq c \lg n \quad \text{con } c < 1$$

Podemos reordenar esta desigualdad:

$$\lg n - 1 \leq c \lg n$$

$$\lg n (1 - c) \leq 1$$

Para que esto sea cierto para cualquier valor de $n$, necesitamos que $c \geq 1$. Sin embargo, la condición indica que $c < 1$, lo cual no se cumple para valores grandes de $n$.

Por lo tanto, la condición de regularidad no se cumple para ninguna constante $c < 1$.

## 2. Verificación del Caso 3 del Teorema Maestro

El caso 3 del Teorema Maestro requiere verificar si:

$$f(n) = \Omega(n^{\log_b a + \epsilon}) \quad \text{para algún } \epsilon > 0$$

Sustituyendo los valores de $a$ y $b$:

- $a = 1$
- $b = 2$

Entonces:

$$\log_b a = \log_2 1 = 0$$

Por lo tanto:

$$n^{\log_b a + \epsilon} = n^{0 + \epsilon} = n^{\epsilon}$$

Para verificar si $f(n) = \lg n$ pertenece a $\Omega(n^{\epsilon})$ para algún $\epsilon > 0$, necesitamos que:

$$\lg n \geq k \cdot n^{\epsilon} \quad \text{para alguna constante } k > 0 \text{ y para } n \text{ suficientemente grande}$$

Sin embargo, $\lg n$ crece mucho más lentamente que cualquier potencia positiva de $n$, es decir, $\lg n$ no puede ser mayor que $n^{\epsilon}$ para valores suficientemente grandes de $n$ y cualquier $\epsilon > 0$. Esto implica que:

$$f(n) \neq \Omega(n^{\epsilon}) \quad \text{para cualquier } \epsilon > 0$$

Por lo tanto, la condición en el caso 3 del Teorema Maestro no se cumple.

# Conclusión

Aunque $f(n/2) < f(n)$, la condición de regularidad no se cumple para ninguna constante $c < 1$. Además, la condición en el caso 3 del Teorema Maestro tampoco se cumple para ningún $\epsilon > 0$.
