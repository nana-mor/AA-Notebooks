La codificación de Huffman es una técnica para la comprensión de datos ampliamente usada y efectiva.

Tomemos un ejemplo para simular un escenario de uso.

Tenemos un fichero con 100000 caracteres, sabemos que aparececen 6 caracteristicas diferentes y la frecuencia de aparición de cada uno de ellos es
 - a = 45

Para poder desarrollar adecuadamente el algoritmo vamos a plantear la pregunta:
- "Como codificar los caracteres para comprimir el espacio ocupado utilizando un código binario?"

Para esto tenemos las posibles soluciones:

- Fija donde usamos los codigos de longitud Fija, para 6 caracteres se nesecitan 3 bits lo que equivale a 300000 bits.

**Fija** | 000 | 001|  010| 011| 100| 101|

- Variable donde la longitud variable en el que los mas frecuentes tienen el codigo mas corto, la restccion es prefijo de otro

**Variable**| 0 |101|100|111|1101|1100|

Esta tacnica se llama codigo prefijo, ya que basta concatenar el codigo de cada uno de los caracteres.

**Ejemplo**

aabacd=0*0*101*0*100*111=001010100111

En la decodificacion al no haber nungun codigo prefigojo de otro no hay ambiguedad por lo que podemos hacer un recorrido de atras hacia delante.
001010100111=aabacd

Dado que **$T$**, el árbol binario que corresponde a una codificación prefijo, es facil averiguar el numero de bits necesarios para codificar el fichero:

- sea $f(c)$ la frecuencia de c en la entrada.
sea $d_r(c)$ la prinfudidad de la hoja $c$ en el árbol **$T$**, entonces el número de bits requerido es: 

$$B(T) = \sum f(c) \cdot d_T(c)$$

$$c \in C$$

$\therefore B(T)$ nos da el coste de **$T$**
