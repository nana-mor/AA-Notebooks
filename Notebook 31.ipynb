**16-1 Código reflejado binario de Gray**
Un código de Gray binario representa una secuencia de enteros no negativos en binario, donde, al pasar de un número entero al siguiente, se cambia exactamente un bit cada vez. El código reflejado binario de Gray genera una secuencia de enteros del 0 al $2^k-1$ para algún entero positivo $k$ según el siguiente método recursivo:

* Para $k = 1$, el código reflejado binario de Gray es $\langle 0, 1 \rangle$.  
* Para $k \geq 2$, primero forma el código reflejado binario de Gray para $k-1$, generando los $2^{k-1}$ enteros del $0$ al $2^{k-1} - 1$. Luego, forma la **reflexión** de esta secuencia, que simplemente es la secuencia en orden inverso. (Es decir, el $j$-ésimo entero en la secuencia se convierte en el $(2^{k-1} - j - 1)$-ésimo entero en la reflexión). Después, suma $2^{k-1}$ a cada uno de los $2^{k-1}$ enteros en la secuencia reflejada. Finalmente, concatena las dos secuencias.

Por ejemplo, para $k = 2$, primero forma el código reflejado binario de Gray $\langle 0, 1 \rangle$ para $k = 1$. Su reflexión es la secuencia $\langle 1, 0 \rangle$. Al sumar $2^{k-1} = 2$ a cada entero en la reflexión, se obtiene $\langle 3, 2 \rangle$. Concatenando ambas secuencias, resulta $\langle 0, 1, 3, 2 \rangle$, o en binario, $\langle 00, 01, 11, 10 \rangle$, donde cada entero difiere de su predecesor en exactamente un bit. Para $k = 3$, la reflexión del código reflejado binario de Gray para $k = 2$ es $\langle 2, 3, 1, 0 \rangle$. Al sumar $2^{k-1} = 4$, se obtiene $\langle 6, 7, 5, 4 \rangle$. Concatenando las secuencias, se produce $\langle 0, 1, 3, 2, 6, 7, 5, 4 \rangle$, o en binario, $\langle 000, 001, 011, 010, 110, 111, 101, 100 \rangle$.  En el código reflejado binario de Gray, solo un bit cambia incluso al pasar del último número entero al primero.

**a.** Indexa los enteros en un código reflejado binario de Gray del $0$ al $2^k - 1$, y considera el $i$-ésimo entero en el código reflejado binario de Gray. Para pasar del entero $(i-1)$ al entero $i$ en el código reflejado binario de Gray, se cambia exactamente un bit. Muestra cómo determinar cuál bit cambia, dado el índice $i$.

**b.** Asumiendo que, dado un número de bit $j$, puedes cambiar el bit $j$ de un entero en tiempo constante, muestra cómo calcular la secuencia completa del código reflejado binario de Gray de $2^k$ números en tiempo $\Theta(2^k)$.
